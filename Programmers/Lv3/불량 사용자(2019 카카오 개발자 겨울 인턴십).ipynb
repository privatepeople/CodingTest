```python
from typing import List, Tuple, Set
import re

def trans_star(stars: re.Match) -> str:
    """banned_id의 각 마스킹된 부분들을 만족하는 정규 패턴으로 변환하는 함수

    Args:
        stars: 패턴 매칭된 문자열

    Returns:
        만족하는 패턴 매칭
    """
    star_cnt = len(stars.group())
    return r'[a-z0-9]{' + f'{star_cnt}' + r'}'

def case_check(pattern_list: List[List[str]], current_set: Set[str], idx: int, result_set: Set[Tuple[str]]) -> None:
    """banned_id마다 패턴 매칭된 유저들을 바탕으로 조합하는 함수

    Args:
        pattern_list: 각 banned_id에 가능한 유저들의 리스트, List[List[str]]
        current_set: 현재까지 조합한 bannded_id 집합, Set[str]
        idx: 현재 pattern_list index, int
        result_set: 최종 조합된 banned_id 집합, Set[Tuple[str]]

    Returns:
        None
    """
    if idx == len(pattern_list):
        result_set.add(tuple(sorted(current_set)))
        return None

    for user in pattern_list[idx]:
        if user not in current_set:
            case_check(pattern_list, current_set | {user}, idx + 1, result_set)

def solution(user_id: List[str], banned_id: List[str]) -> int:
    answer = 0
    pattern_list = []
    p = re.compile(pattern=r"\*+")
    n = len(banned_id)
    result_set = set()

    for banned in banned_id:
        banned_p = p.sub(repl=trans_star, string=banned)
        banned_p = '^' + banned_p + '$'
        matching_list = []
        banned_p = re.compile(pattern=banned_p)

        for user in user_id:
            result = banned_p.match(string=user)
            if result is not None:
                matching_list.append(result.group())

        pattern_list.append(matching_list)

    case_check(pattern_list, set(), 0, result_set)
    answer = len(result_set)
    return answer

# TestCase
print(solution(["frodo", "fradi", "crodo", "abc123", "frodoc"], ["fr*d*", "abc1**"]) == 2)
print(solution(["frodo", "fradi", "crodo", "abc123", "frodoc"], ["*rodo", "*rodo", "******"]) == 2)
print(solution(["frodo", "fradi", "crodo", "abc123", "frodoc"], ["fr*d*", "*rodo", "******", "******"]) == 3)
```
