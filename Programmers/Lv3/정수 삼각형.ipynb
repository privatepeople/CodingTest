```python
from typing import List

def solution(triangle: List[List[int]]) -> int:
    n = len(triangle)
    if n == 1:
        return triangle[0][0]

    # 상향식 방식으로 최댓값 계산
    for i in range(n - 1, 0, -1):
        k = len(triangle[i])
        for j in range(k - 1):
            triangle[i - 1][j] += max(triangle[i][j], triangle[i][j + 1])

    return triangle[0][0]

# TestCase
print(solution([[7], [3, 8], [8, 1, 0], [2, 7, 4, 4], [4, 5, 2, 6, 5]]) == 30)
```
