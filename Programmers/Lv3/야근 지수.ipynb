```python
from typing import List
from collections import Counter
from operator import itemgetter

def solution(n: int, works: List[int]) -> int:
    if sum(works) <= n:
        return 0

    work_remain_list = sorted(Counter(works).items(), key=itemgetter(0))

    while n:
        times, amount = work_remain_list.pop()
        if n >= amount :
            if work_remain_list and work_remain_list[-1][0] == times - 1:
                work_remain_list[-1] = (work_remain_list[-1][0], work_remain_list[-1][1] + amount)
            else:
                work_remain_list.append((times - 1, amount))
            n -= amount
        else:
            if work_remain_list and work_remain_list[-1][0] == times - 1:
                work_remain_list[-1] = (work_remain_list[-1][0], work_remain_list[-1][1] + n)
            else:
                work_remain_list.append((times - 1, n))
            work_remain_list.append((times, amount - n))
            n = 0

    return sum(map(lambda x: x[1] * (x[0]**2), work_remain_list))

# TestCase
print(solution(4, [4, 3, 3]) == 12)
print(solution(1, [2, 1, 2]) == 6)
print(solution(3, [1, 1]) == 0)
print(solution(10, [10, 10, 1]) == 51)
print(solution(999, [800, 100, 55, 45]) == 1)
print(solution(99, [2, 15, 22, 55, 55]) == 580)
```
