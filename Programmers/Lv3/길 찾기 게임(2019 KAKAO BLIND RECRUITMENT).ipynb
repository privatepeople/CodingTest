```python
from typing import List, Optional
from sys import setrecursionlimit
from operator import itemgetter
setrecursionlimit(10**4)

class Node:
    """트리 자료구조의 노드를 나타내는 클래스"""
    def __init__(self, x: int, y: int, number: int, left=None, right=None) -> None:
        """클래스 생성자"""
        self.x = x
        self.y = y
        self.number = number
        self.left, self.right = left, right


class Tree:
    """트리 자료구조를 나타내는 클래스"""
    def __init__(self) -> None:
        """클래스 생성자"""
        self.root = None

    def insert(self, root: Optional[Node], new_node: Node) -> None:
        """트리 자료구조에 노드를 삽입하는 메소드"""
        if self.root is None:
            self.root = new_node
        else:
            if root.x < new_node.x:
                if root.right is None:
                    root.right = new_node
                else:
                    self.insert(root.right, new_node)
            else:
                if root.left is None:
                    root.left = new_node
                else:
                    self.insert(root.left, new_node)
        return None

    def preorder(self, node: Node, result: Optional[List[int]] = None) -> List[int]:
        """트리 자료구조를 전위순회하는 메소드"""
        if result is None:
            result = []

        result.append(node.number)
        if node.left is not None:
            result = self.preorder(node.left, result)
        if node.right is not None:
            result = self.preorder(node.right, result)
        return result

    def postorder(self, node: Node, result: Optional[List[int]] = None) -> List[int]:
        """트리 자료구조를 후위순회하는 메소드"""
        if result is None:
            result = []

        if node.left is not None:
            result = self.postorder(node.left, result)
        if node.right is not None:
            result = self.postorder(node.right, result)
        result.append(node.number)
        return result


def solution(nodeinfo: List[List[int]]) -> List[List[int]]:
    nodeinfo = [(x, y, number) for number, (x, y) in enumerate(nodeinfo, start=1)]
    nodeinfo.sort(key=itemgetter(1), reverse=True)
    tree = Tree()

    for x, y, number in nodeinfo:
        tree.insert(tree.root, Node(x, y, number))

    return [tree.preorder(tree.root), tree.postorder(tree.root)]

# TestCase
print(solution([[5, 3], [11, 5], [13, 3], [3, 5], [6, 1], [1, 3], [8, 6], [7, 2], [2, 2]]) == [[7, 4, 6, 9, 1, 8, 5, 2, 3], [9, 6, 5, 8, 1, 4, 3, 2, 7]])
```
