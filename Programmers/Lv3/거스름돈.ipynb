```python
from typing import List

def solution(n: int, money: List[int]) -> int:
    money_list = [1] + [0] * n

    for coin in sorted(money):
        for price in range(coin, n + 1):
            money_list[price] += money_list[price - coin]

    return money_list[n] % 1000000007

# TestCase
print(solution(5, [1, 2, 5]) == 4)
```
