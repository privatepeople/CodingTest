```python
from typing import List
from collections import deque

def solution(stones: List[int], k: int) -> int:
    window = []
    Qi = deque()
    N = len(stones)

    for i in range(k):
        while Qi and stones[i] >= stones[Qi[-1]]:
            Qi.pop()

        Qi.append(i)

    for i in range(k, N):
        window.append(stones[Qi[0]])

        while Qi and Qi[0] <= i - k:
            Qi.popleft()

        while Qi and stones[i] >= stones[Qi[-1]]:
            Qi.pop()

        Qi.append(i)

    window.append(stones[Qi[0]])
    return min(window)

# TestCase
print(solution([2, 4, 5, 3, 2, 1, 4, 2, 5, 1], 3) == 3)
```
