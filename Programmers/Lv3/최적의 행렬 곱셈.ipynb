```python
from typing import List
from math import inf

def solution(matrix_sizes: List[List[int]]) -> int:
    n = len(matrix_sizes)
    matrix_operation = [[inf for j in range(n)] for i in range(n)]
    for i in range(n):
        matrix_operation[i][i] = 0

    for diagonal in range(1, n):
        for i in range(n - diagonal):
            j = i + diagonal
            for k in range(i, j):
                value = matrix_operation[i][k] + matrix_operation[k + 1][j]
                value += matrix_sizes[i][0] * matrix_sizes[k][1] * matrix_sizes[j][1]
                matrix_operation[i][j] = min(matrix_operation[i][j], value)

    answer = matrix_operation[0][-1]
    return answer

# TestCase
print(solution([[5, 3], [3, 10], [10, 6]]) == 270)
```
