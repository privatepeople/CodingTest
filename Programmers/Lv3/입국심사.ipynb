```python
from typing import List

def solution(n: int, times: List[int]) -> int:
    start, end = 1, times[-1] * n

    while start < end:
        mid = (start + (end - start) // 2)
        total = sum(map(lambda t: mid // t, times)) # mid초 동안 n명의 심사관들이 심사한 사람의 수

        if total >= n:
            end = mid
        else:
            start = mid + 1

    answer = start
    return answer

# TestCase
print(solution(6, [7, 10]) == 28)
```
