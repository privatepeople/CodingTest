```python
from typing import List
from collections import defaultdict
import heapq

def solution(n: int, edge: List[List[int]]) -> int:
    node_cost_list = [None] * (n + 1)
    graph = defaultdict(list)
    for v1, v2 in edge:
        graph[v1].append(v2)
        graph[v2].append(v1)

    hq = [(0, 1)]
    while hq:
        cost, v = heapq.heappop(hq)
        if node_cost_list[v] is None:
            node_cost_list[v] = cost
            for node in graph[v]:
                if node_cost_list[node] is None:
                    heapq.heappush(hq, (cost + 1, node))

    answer, max_cost = 0, -1
    for cost in node_cost_list[1:]:
        if cost > max_cost:
            max_cost = cost
            answer = 1
        elif cost == max_cost:
            answer += 1

    return answer

# TestCase
print(solution(6, [[3, 6], [4, 3], [3, 2], [1, 3], [1, 2], [2, 4], [5, 2]]) == 3)
```
