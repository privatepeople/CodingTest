```python
from typing import List
from operator import itemgetter

def solution(routes: List[List[int]]) -> int:
    routes.sort(key=itemgetter(1, 0))
    end_time = routes[0][1]
    answer = 1

    for start, end in routes:
        if start > end_time:
            answer += 1
            end_time = end

    return answer

# TestCase
print(solution([[-20, -15], [-14, -5], [-18, -13], [-5, -3]]) == 2)
```
