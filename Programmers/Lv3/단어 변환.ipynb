```python
from typing import List
from collections import deque

def str_comparison(s1: str, s2: str) -> bool:
    """두 문자열이 한 글자만 다른지 확인하는 함수"""
    cnt = 0

    for chr1, chr2 in zip(s1, s2):
        if chr1 != chr2:
            cnt += 1
        if cnt >= 2:
            return False

    return True if cnt == 1 else False


def solution(begin: str, target: str, words: List[str]) -> int:
    word_set = set(words)
    if (begin == target) or (target not in word_set):
        return 0

    dq = deque([(begin, 0)])
    while dq:
        current_word, cnt = dq.popleft()
        next_str = tuple(word for word in word_set if str_comparison(current_word, word))

        for word in next_str:
            if word == target:
                return cnt + 1
            else:
                word_set.remove(word)
                dq.append((word, cnt + 1))

    return 0

# TestCase
print(solution("hit", "cog", ["hot", "dot", "dog", "lot", "log", "cog"]) == 4)
print(solution("hit", "cog", ["hot", "dot", "dog", "lot", "log"]) == 0)
print(solution("hit", "wow", ["hot", "dog", "dot", "wow"]) == 0)
print(solution("aoa", "aof", ["aob", "aoc", "aod", "aof", "aoe"]) == 1)
print(solution("hit", "bak", ["hot", "dot", "dog", "lot", "log", "bak"]) == 0)
```
