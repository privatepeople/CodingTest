```python
from typing import List
from operator import itemgetter

def disjoint_find(x: int, parents: List[int]) -> int:
    """루트노드를 찾는 함수

    x : 노드 번호
    parents : 각 노드의 루트 노드들을 모아놓은 리스트
    """
    if x != parents[x]:
        parents[x] = disjoint_find(parents[x], parents)
    return parents[x]


def disjoint_union(x: int, y: int, parents: List[int], rank: List[int]) -> None:
    """루트노드를 하나로 합치는 함수

    x : 노드 번호
    y : 노드 번호
    parents : 각 노드의 루트 노드들을 모아놓은 리스트
    rank : 트리의 높이
    """
    x, y = disjoint_find(x, parents), disjoint_find(y, parents)
    if rank[x] > rank[y]:
        parents[y] = x
    else:
        parents[x] = y
    if rank[x] == rank[y]:
        rank[y] += 1
    return None


def solution(n: int, costs: List[List[int]]) -> int:
    parents = list(range(n))
    rank = [0] * n
    costs.sort(key=itemgetter(2))
    answer = 0

    for start, end, cost in costs:
        x, y = disjoint_find(start, parents), disjoint_find(end, parents)
        if x != y:
            disjoint_union(x, y, parents, rank)
            answer += cost
            n -= 1
        if n == 1:
            break

    return answer

# TestCase
print(solution(4, [[0, 1, 1], [0, 2, 2], [1, 2, 5], [1, 3, 1], [2, 3, 8]]) == 4)
```
