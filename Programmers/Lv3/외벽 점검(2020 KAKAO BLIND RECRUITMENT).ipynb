```python
from typing import List
from itertools import permutations

def solution(n: int, weak: List[int], dist: List[int]) -> int:
    weak_length = len(weak)
    dist.sort(reverse=True)

    # 원형 외벽을 일자로 만들기 위해 두 배로 늘림
    weak.extend([i + n for i in weak])

    answer = len(dist) + 1  # 친구 수 최댓값 + 1로 초기화

    # 가능한 모든 친구의 순열을 계산
    for start in range(weak_length):
        for friends in permutations(dist):
            count = 1  # 첫 번째 친구 사용
            position = weak[start] + friends[count - 1]
            for index in range(start, start + weak_length):
                if position < weak[index]:
                    count += 1  # 새로운 친구 추가
                    if count > len(friends):
                        break
                    position = weak[index] + friends[count - 1]
            answer = min(answer, count)

    if answer > len(dist):
        return -1

    return answer

# TestCase
print(solution(12, [1, 5, 6, 10], [1, 2, 3, 4]) == 2)
print(solution(12, [1, 3, 4, 9, 10], [3, 5, 7]) == 1)
print(solution(30, [0, 3, 11, 21], [10, 4]) == 2)
print(solution(50, [1], [6]) == 1)
print(solution(200, [0, 10, 50, 80, 120, 160], [1, 10, 5, 40, 30]) == 3)
print(solution(200, [0, 100], [1, 1]) == 2)
print(solution(6, [1, 2, 4, 5], [1, 1]) == 2)
print(solution(12, [0, 10], [1, 2]) == 1)
```
