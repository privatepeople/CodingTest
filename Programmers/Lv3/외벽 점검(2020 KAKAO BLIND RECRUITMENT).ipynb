```python
from typing import List
from collections import deque

def solution(n: int, weak: List[int], dist: List[int]) -> int:
    dist.sort()
    dq = deque([(0, weak, len(dist) - 1)])

    while dq:
        num, remain_weak, dist_idx = dq.popleft()
        moving = dist[dist_idx]
        num += 1

        for idx, start in enumerate(remain_weak):
            end = start + moving
            if end < n:
                if idx + 1 < len(remain_weak) and remain_weak[idx + 1] > end:
                    continue

                next_weak = tuple(i for i in remain_weak if i < start or i > end)
            else:
                remain = end % n
                next_weak = tuple(i for i in remain_weak if remain < i < start)

            if not next_weak:
                return num
            elif dist_idx > 0:
                dq.append((num, next_weak, dist_idx - 1))

    return -1

# TestCase
print(solution(12, [1, 5, 6, 10], [1, 2, 3, 4]) == 2)
print(solution(12, [1, 3, 4, 9, 10], [3, 5, 7]) == 1)
print(solution(30, [0, 3, 11, 21], [10, 4]) == 2)
print(solution(50, [1], [6]) == 1)
print(solution(200, [0, 10, 50, 80, 120, 160], [1, 10, 5, 40, 30]) == 3)
print(solution(200, [0, 100], [1, 1]) == 2)
print(solution(6, [1, 2, 4, 5], [1, 1]) == 2)
print(solution(12, [0, 10], [1, 2]) == 1)
```
