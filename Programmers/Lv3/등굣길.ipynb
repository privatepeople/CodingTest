```python
from typing import List, Set, Dict, Tuple

def cal_minimum(x: int, y: int, m: int, n: int, puddle_set: Set[Tuple[int]], map_dict: Dict[Tuple[int], int]) -> int:
    """(x, y)위치까지의 최단경로를 반환하는 함수

    x : 현재 x좌표
    y : 현재 y좌표
    m : 격자의 가로길이
    n : 격자의 세로길이
    puddle_set : 물웅덩이의 위치들을 모아놓은 집합
    map_dict : 키 = (x, y)좌표, 값 = 최소 이동 횟수로 구성된 딕셔너리
    """
    if (x, y) in map_dict:
        return map_dict[(x, y)]

    x_, y_ = x - 1, y - 1
    if 1 <= x_ <= m and 1 <= y_ <= n:
        map_dict[(x, y)] = cal_minimum(x_, y, m, n, puddle_set, map_dict) + cal_minimum(x, y_, m, n, puddle_set, map_dict)
    elif 1 <= x_ <= m:
        map_dict[(x, y)] = cal_minimum(x_, y, m, n, puddle_set, map_dict)
    elif 1 <= y_ <= n:
        map_dict[(x, y)] = cal_minimum(x, y_, m, n, puddle_set, map_dict)
    return map_dict[(x, y)]


def solution(m: int, n: int, puddles: List[List[int]]) -> int:
    puddle_set = set((x, y) for x, y in puddles)
    map_dict = {(x, y): 0 for x, y in puddles}
    for a, b in [(1, 2), (2, 1)]:
        if (a, b) not in puddle_set:
            map_dict[(a, b)] = 1

    cal_minimum(m, n, m, n, puddle_set, map_dict)
    return map_dict[(m, n)] % 1000000007

# TestCase
print(solution(4, 3, [[2, 2]]) == 4)
```
