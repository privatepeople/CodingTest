```python
from typing import List

def solution(n: int, s: int) -> List[int]:
    if s < n:
        return [-1]

    q, r = divmod(s, n)
    return [q] * (n - r) + [q + 1] * r

# TestCase
print(solution(2, 9) == [4, 5])
print(solution(2, 1) == [-1])
print(solution(2, 8) == [4, 4])
```
