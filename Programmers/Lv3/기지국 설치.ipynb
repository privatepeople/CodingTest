```python
from typing import List
from math import ceil

def solution(n: int, stations: List[int], w: int) -> int:
    answer = 0
    loc = 1

    for i in stations:
        answer += ceil(max(i - w - loc, 0) / (w * 2 + 1))
        loc = (i + w + 1)

    answer += ceil(max(n - loc + 1, 0) / (w * 2 + 1))
    return answer

# TestCase
print(solution(11, [4, 11], 1) == 3)
print(solution(16, [9], 2) == 3)
```
