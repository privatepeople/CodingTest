```python
from typing import List
from collections import defaultdict

def solution(tickets: List[List[str]]) -> List[str]:
    ticket_dict = defaultdict(list)
    for start, end in tickets:
        ticket_dict[start].append(end)
    for start in ticket_dict:
        ticket_dict[start].sort(reverse=True)

    stack = ["ICN"]
    path = []
    while stack:
        top = stack[-1]
        # 더 이상 방문할 수 없으면은 stack에서 꺼내 path에 추가
        if (top not in ticket_dict) or (not ticket_dict[top]):
            path.append(stack.pop())
        else:
            stack.append(ticket_dict[top].pop())

    return path[::-1]

# TestCase
print(solution([["ICN", "JFK"], ["HND", "IAD"], ["JFK", "HND"]]) == ["ICN", "JFK", "HND", "IAD"])
print(solution([["ICN", "SFO"], ["ICN", "ATL"], ["SFO", "ATL"], ["ATL", "ICN"], ["ATL", "SFO"]]) == ["ICN", "ATL", "ICN", "SFO", "ATL", "SFO"])
print(solution([["ICN", "COO"], ["ICN", "BOO"], ["COO", "ICN"]]) == ["ICN", "COO", "ICN", "BOO"])
print(solution([["ICN", "SFO"], ["SFO", "ICN"], ["ICN", "SFO"], ["SFO", "QRE"]]) == ["ICN", "SFO", "ICN", "SFO", "QRE"])
print(solution([["ICN", "AFO"], ["ICN", "PTL"], ["AFO", "PTL"], ["PTL", "ICN"], ["PTL", "AFO"]]) == ["ICN", "AFO", "PTL", "ICN", "PTL", "AFO"])
```
