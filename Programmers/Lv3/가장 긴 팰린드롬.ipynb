```python
def ispalindrome(left: int, right: int, s_len: int, s: str) -> str:
    """시작 위치로부터 길이가 최대인 팰린드롬을 반환하는 함수

    left : 문자열 s의 왼쪽 인덱스 위치
    right : 문자열 s의 오른쪽 인덱스 위치
    s_len : 문자열 s의 길이
    s: 문자열
    """
    while 0 <= left and right < s_len and s[left] == s[right]:
        left, right = left - 1, right + 1
    return s[left + 1:right]


def solution(s: str) -> int:
    n = len(s)
    answer = ''
    if n < 2 or s == s[::-1]:
        return n

    for i in range(n - 1):
        s1, s2 = ispalindrome(i, i + 1, n, s), ispalindrome(i, i + 2, n, s)
        answer = max((answer, s1, s2), key=len)

    return len(answer)

# TestCase
print(solution("abcdcba") == 7)
print(solution("abacde") == 3)
```
