```python
from typing import List
from collections import defaultdict
from operator import itemgetter

def solution(genres: List[str], plays: List[int]) -> List[int]:
    genre_dict = defaultdict(int)
    genre_list = defaultdict(list)
    answer = []

    for idx, (genre, play) in enumerate(zip(genres, plays)):
        genre_dict[genre] += play
        genre_list[genre].append((-idx, play))
    for genre in genre_list:
        genre_list[genre].sort(key=itemgetter(1, 0), reverse=True)

    for genre, cnt in sorted(genre_dict.items(), key=itemgetter(1), reverse=True):
        genre_top2 = genre_list[genre][:min(2, len(genre_list[genre]))]
        answer.extend([-idx for idx, play in genre_top2])

    return answer

# TestCase
print(solution(["classic", "pop", "classic", "classic", "pop"], [500, 600, 150, 800, 2500]) == [4, 1, 3, 0])
```
