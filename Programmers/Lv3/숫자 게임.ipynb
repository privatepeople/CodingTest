```python
from typing import List

def solution(A: List[int], B: List[int]) -> int:
    A.sort()
    B.sort()
    answer = 0
    j = 0

    for i in range(len(B)):
        if A[j] < B[i]:
            answer += 1
            j += 1

    return answer

# TestCase
print(solution([5, 1, 3, 7], [2, 2, 6, 8]) == 3)
print(solution([2, 2, 2, 2], [1, 1, 1, 1]) == 0)
```
