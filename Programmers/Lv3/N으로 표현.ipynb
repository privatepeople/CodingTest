```python
def solution(N: int, number: int) -> int:
    dp = {i: set() for i in range(1, 9)} # 사용횟수: 만족하는 자연수 집합으로 구성된 딕셔너리

    for i in range(1, 9):
        cnt_set = dp[i]
        cnt_set.add(int(str(N) * i)) # N, NN, NNN, ...
        for j in range(1, i):
            for k in dp[j]:
                for l in dp[i - j]:
                    cnt_set.add(k + l)
                    cnt_set.add(k * l)
                    if k >= l:
                        cnt_set.add(k - l)
                    if l and k:
                        cnt_set.add(k // l)

        if number in cnt_set:
            return i

    return -1

# TestCase
print(solution(5, 12) == 4)
print(solution(5, 31168) == -1)
print(solution(5, 137) == 6)
print(solution(5, 138) == 6)
print(solution(3, 4) ==3)
print(solution(5, 24) == 4)
print(solution(5, 26) == 4)
print(solution(4, 17) == 4)
print(solution(5, 5) == 1)
print(solution(8, 5800) == 8)
print(solution(8, 53) == 5)
print(solution(6, 5) == 3)
print(solution(1, 1121) == 7)
print(solution(2, 22222) == 5)
print(solution(5, 3600) == 6)
print(solution(1, 37) == 6)
```
