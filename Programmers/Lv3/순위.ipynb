```python
from typing import List

def solution(n: int, results: List[List[int]]) -> int:
    graph = [[0 for j in range(n)] for i in range(n)]
    # A가 승자, B는 패자일때 승리했으면 1, 패배했으면 -1
    for A, B in results:
        graph[A - 1][B - 1], graph[B - 1][A - 1] = 1, -1

    # Floyd-Warshall Algorithm을 활용하여 A -> B & B -> C이면 A -> C임을 계산
    for k in range(n):
        for i in range(n):
            if i != k:
                for j in range(n):
                    if j != i and j != k:
                        if graph[i][k] == 1 and graph[k][j] == 1:
                            graph[i][j] = 1
                            graph[j][i] = -1

    # 주 대각성분을 뺀 나머지가 0이 아니면 모든 승부기록을 알 수 있음
    return len([row for row in graph if row.count(0) == 1])

# TestCase
print(solution(5, [[4, 3], [4, 2], [3, 2], [1, 2], [2, 5]]) == 2)
```
