```python
def solution(my_string: str) -> str:
    answer = []
    char_set = set()

    for char in my_string:
        if char not in char_set:
            answer.append(char)
            char_set.add(char)

    return ''.join(answer)

# TestCase
print(solution("people") == "peol")
print(solution("We are the world") == "We arthwold")
```
