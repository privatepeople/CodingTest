```python
from typing import List

def solution(str_list: List[str], ex: str) -> str:
    answer = []

    for s in str_list:
        if ex not in s:
            answer.append(s)

    return ''.join(answer)

# TestCase
print(solution(["abc", "def", "ghi"], "ef") == "abcghi")
print(solution(["abc", "bbc", "cbc"], "c") == "")
```
