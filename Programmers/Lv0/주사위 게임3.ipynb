```python
from collections import Counter

def solution(a: int, b: int, c: int, d: int) -> int:
    dice_dict = Counter([a, b, c, d]).most_common()

    if dice_dict[0][1] == 4:
        return dice_dict[0][0] * 1111
    elif dice_dict[0][1] == 3:
        return (dice_dict[0][0] * 10 + dice_dict[1][0])**2
    elif dice_dict[0][1] == 2 and dice_dict[1][1] == 2:
        return (dice_dict[0][0] + dice_dict[1][0]) * abs(dice_dict[0][0] - dice_dict[1][0])
    elif dice_dict[0][1] == 2:
        return dice_dict[1][0] * dice_dict[2][0]
    else:
        return min(a, b, c, d)

# TestCase
print(solution(2, 2, 2, 2) == 2222)
print(solution(4, 1, 4, 4) == 1681)
print(solution(6, 3, 3, 6) == 27)
print(solution(2, 5, 2, 6) == 30)
print(solution(6, 4, 2, 5) == 2)
```
