```python
from math import gcd

def solution(a: int, b: int) -> int:
    b //= gcd(a, b)
    while not (b & 1):
        b >>= 1
    while True:
        q, r = divmod(b, 5)
        if r:
            break
        else:
            b = q

    return 1 if b == 1 else 2

# TestCase
print(solution(7, 20) == 1)
print(solution(11, 22) == 1)
print(solution(12, 21) == 2)
```
