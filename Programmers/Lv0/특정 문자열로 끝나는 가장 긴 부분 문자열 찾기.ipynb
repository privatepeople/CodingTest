```python
def solution(myString: str, pat: str) -> str:
    last_index = myString.rindex(pat) + len(pat)
    return myString[:last_index]

# TestCase
print(solution("AbCdEFG", "dE") == "AbCdE")
print(solution("AAAAaaaa", "a") == "AAAAaaaa")
```
