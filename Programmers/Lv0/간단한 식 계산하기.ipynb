```python
def solution(binomial: str) -> int:
    equal_list = binomial.split()

    if equal_list[1] == '+':
        return int(equal_list[0]) + int(equal_list[2])
    elif equal_list[1] == '-':
        return int(equal_list[0]) - int(equal_list[2])
    else:
        return int(equal_list[0]) * int(equal_list[2])

# TestCase
print(solution("43 + 12") == 55)
print(solution("0 - 7777") == -7777)
print(solution("40000 * 40000") == 1600000000)
```
