```python
def solution(my_string: str, s: int, e: int) -> str:
    e += 1
    return my_string[:s] + my_string[s:e][::-1] + my_string[e:]

# TestCase
print(solution("Progra21Sremm3", 6, 12) == "ProgrammerS123")
print(solution("Stanley1yelnatS", 4, 10) == "Stanley1yelnatS")
```
