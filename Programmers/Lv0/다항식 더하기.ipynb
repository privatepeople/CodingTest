```python
def solution(polynomial: str) -> str:
    x_coff = constant = 0
    polynomial_list = polynomial.split(' + ')
    answer = ''

    for term in polynomial_list:
        if 'x' in term:
            x_coff += int(term[:-1]) if term != 'x' else 1
        else:
            constant += int(term)

    if x_coff > 1:
        answer += str(x_coff)
    if x_coff >= 1:
        answer += 'x'
    if constant > 0:
        if answer:
            answer += f' + {constant}'
        else:
            answer = str(constant)

    return answer

# TestCase
print(solution("3x + 7 + x") == "4x + 7")
print(solution("x + x + x") == "3x")
```
