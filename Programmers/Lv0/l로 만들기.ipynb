```python
def solution(myString: str) -> str:
    # ord(문자) -> 아스키코드 숫자, chr(아스키코드 숫자) -> 문자
    alpha_trans_dict = {chr(i): 'l' for i in range(97, 108)}
    return myString.translate(str.maketrans(alpha_trans_dict))

# TestCase
print(solution("abcdevwxyz") == "lllllvwxyz")
print(solution("jjnnllkkmm") == "llnnllllmm")
```
