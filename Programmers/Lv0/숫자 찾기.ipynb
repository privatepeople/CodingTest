문제 출처 : https://school.programmers.co.kr/learn/courses/30/lessons/120904

```python
def solution(num: int, k: int) -> int:
    num, k = str(num), str(k)

    for idx, char in enumerate(num, start=1):
        if char == k:
            return idx

    return -1

# TestCase
print(solution(29183, 1) == 3)
print(solution(232443, 4) == 4)
print(solution(123456, 7) == -1)
```
