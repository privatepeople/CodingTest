```python
from typing import Iterator

def sieve_eratosthenes(n: int) -> Iterator[int]:
    """n이하 자연수 중에서 소수들을 오름차순으로 반환하는 제너레이터"""
    prime_set = set(range(2, n + 1))

    for i in range(2, int(n**0.5) + 1):
        if i in prime_set:
            prime_set -= set(range(i * 2, n + 1, i))

    yield from sorted(prime_set)

def solution(n: int) -> int:
    # n개의 양의 정수 중에서 에라토스테네스의 체로 n이하의 소수들의 집합의 크기와 1을 뺀 나머지
    return n - len(list(sieve_eratosthenes(n))) - 1

# TestCase
print(solution(10) == 5)
print(solution(15) == 8)
```
