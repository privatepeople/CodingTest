```python
from typing import List
from operator import itemgetter

def solution(rank: List[int], attendance: List[bool]) -> int:
    rank_list = [(i, ranked) for i, (ranked, attend) in enumerate(zip(rank, attendance)) if attend]
    rank_list.sort(key=itemgetter(1))

    return 10000 * rank_list[0][0] + 100 * rank_list[1][0] + rank_list[2][0]

# TestCase
print(solution([3, 7, 2, 5, 4, 6, 1], [False, True, True, True, True, False, False]) == 20403)
print(solution([1, 2, 3], [True, True, True]) == 102)
print(solution([6, 1, 5, 2, 3, 4], [True, False, True, False, False, True]) == 	50200)
```
