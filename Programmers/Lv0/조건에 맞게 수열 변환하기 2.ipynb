```python
from typing import List

def trans_num(num: int) -> int:
    """수열을 조건에 맞게 변환하는 함수"""
    if num >= 50 and (num ^ 1) & 1:
        return num >> 1
    elif num < 50 and num & 1:
        return (num << 1) + 1
    else:
        return num

def solution(arr: List[int]) -> int:
    answer = 0

    while True:
        next_arr = list(map(trans_num, arr))

        if arr == next_arr:
            return answer
        else:
            arr = next_arr
            answer += 1

# TestCase
print(solution([1, 2, 3, 100, 99, 98]) == 5)
```
