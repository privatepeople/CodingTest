```python
from typing import List

def solution(score: List[List[int]]) -> List[int]:
    sum_list = sorted([sum(subject_list) for subject_list in score], reverse=True)
    rank_dict = {}
    for idx, scores in enumerate(sum_list, start=1):
        if scores not in rank_dict:
            rank_dict[scores] = idx

    return [rank_dict[sum(total_score)] for total_score in score]

# TestCase
print(solution([[80, 70], [90, 50], [40, 70], [50, 80]]) == [1, 2, 4, 3])
print(solution([[80, 70], [70, 80], [30, 50], [90, 100], [100, 90], [100, 100], [10, 30]]) == [4, 4, 6, 2, 2, 1, 7])
```
