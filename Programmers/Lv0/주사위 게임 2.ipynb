문제 출처 : https://school.programmers.co.kr/learn/courses/30/lessons/181930

```python
def solution(a: int, b: int, c: int) -> int:
    n = len(set([a, b, c]))

    if n == 3:
        return a + b + c
    elif n == 2:
        return (a + b + c) * ((a**2) + (b**2) + (c**2))
    else:
        return (a + b + c) * ((a**2) + (b**2) + (c**2)) * ((a**3) + (b**3) + (c**3))

# TestCase
print(solution(2, 6, 1) == 9)
print(solution(5, 3, 3) == 473)
print(solution(4, 4, 4) == 110592)
```
