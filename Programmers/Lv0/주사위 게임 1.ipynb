```python
def solution(a: int, b: int) -> int:
    if a & 1 and b & 1:
        return (a**2) + (b**2)
    elif a & 1 or b & 1:
        return 2 * (a + b)
    else:
        return abs(a - b)

# TestCase
print(solution(3, 5) == 34)
print(solution(6, 1) == 14)
print(solution(2, 4) == 2)
```
