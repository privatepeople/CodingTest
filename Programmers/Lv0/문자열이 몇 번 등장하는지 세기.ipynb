```python
def solution(myString: str, pat: str) -> int:
    answer = 0
    len_pat = len(pat)

    for i in range(len(myString) - len_pat + 1):
        if myString[i:i + len_pat] == pat:
            answer += 1

    return answer

# TestCase
print(solution("banana", "ana") == 2)
print(solution("aaaa", "aa") == 3)
```
