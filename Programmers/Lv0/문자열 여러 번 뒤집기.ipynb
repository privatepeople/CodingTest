```python
from typing import List

def solution(my_string: str, queries: List[List[int]]) -> str:
    for start, end in queries:
        end += 1
        my_string = my_string[:start] + my_string[start:end][::-1] + my_string[end:]

    return my_string

# TestCase
print(solution("rermgorpsam", [[2, 3], [0, 7], [5, 9], [6, 10]]) == "programmers")
```
