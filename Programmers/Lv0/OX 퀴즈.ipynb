```python
from typing import List

def solution(quiz: List[str]) -> List[str]:
    answer = []

    for q in quiz:
        L, R = q.split(' = ')
        a, op, b = L.split()
        if op == '+':
            result = 'O' if int(a) + int(b) == int(R) else 'X'
        else:
            result = 'O' if int(a) - int(b) == int(R) else 'X'
        answer.append(result)

    return answer

# TestCase
print(solution(["3 - 4 = -3", "5 + 6 = 11"]) == ["X", "O"])
print(solution(["19 - 6 = 13", "5 + 66 = 71", "5 - 15 = 63", "3 - 1 = 2"]) == ["O", "O", "X", "O"])
```
