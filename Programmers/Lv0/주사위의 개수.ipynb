```python
from typing import List

def solution(box: List[int], n: int) -> int:
    x, y, z = box
    return (x // n) * (y // n) * (z // n)

# TestCase
print(solution([1, 1, 1], 1) == 1)
print(solution([10, 8, 6], 3) == 12)
```
