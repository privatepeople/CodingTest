```python
def solution(str1: str, str2: str) -> str:
    answer = []

    for a, b in zip(str1, str2):
        answer.append(a + b)

    return ''.join(answer)

# TestCase
print(solution("aaaaa", "bbbbb") == "ababababab")
```
