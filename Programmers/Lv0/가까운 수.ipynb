```python
from typing import List

def solution(array: List[int], n: int) -> int:
    return min(array, key=lambda x: (abs(n - x), x))

# TestCase
print(solution([3, 10, 28], 20) == 28)
print(solution([10, 11, 12], 13) == 12)
```
