```python
from typing import List
from collections import deque

def solution(l: int, r: int) -> List[int]:
    answer = [5] if l <= 5 <= r else []
    dq = deque([5])

    while dq:
        num = dq.popleft() * 10
        for i in [num, num + 5]:
            if i <= r:
                dq.append(i)
                if i >= l:
                    answer.append(i)

    return answer if answer else [-1]

# TestCase
print(solution(5, 555) == [5, 50, 55, 500, 505, 550, 555])
print(solution(10, 20) == [-1])
```
