```python
from typing import List

def solution(numLog: List[int]) -> str:
    c_dict = {1: 'w', -1: 's', 10: 'd', -10: 'a'}
    answer = []

    for i in range(1, len(numLog)):
        answer.append(c_dict[numLog[i] - numLog[i-1]])

    return ''.join(answer)

# TestCase
print(solution([0, 1, 0, 10, 0, 1, 0, 10, 0, -1, -2, -1]) == "wsdawsdassw")
```
