```python
from typing import List, Iterator

def sieve_eratosthenes(n: int) -> Iterator[int]:
    """n이하 자연수 중에서 소수들을 오름차순으로 반환하는 제너레이터"""
    prime_set = set(range(2, n + 1))

    for i in range(2, int(n**0.5) + 1):
        if i in prime_set:
            prime_set -= set(range(i * 2, n + 1, i))

    yield from sorted(prime_set)

def solution(n: int) -> List[int]:
    return [i for i in sieve_eratosthenes(n) if not (n % i)]

# TestCase
print(solution(12) == [2, 3])
print(solution(17) == [17])
print(solution(420) == [2, 3, 5, 7])
```
