```python
from typing import List

def solution(my_strings: List[str], parts: List[List[int]]) -> str:
    answer = []

    for idx, word in enumerate(my_strings):
        start, end = parts[idx][0], parts[idx][1] + 1
        answer.append(word[start:end])

    return ''.join(answer)

# TestCase
print(solution(["progressive", "hamburger", "hammer", "ahocorasick"], [[0, 4], [1, 2], [3, 5], [7, 7]]) == "programmers")
```
