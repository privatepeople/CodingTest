```python
from typing import List
import re

def solution(babbling: List[str]) -> int:
    regex = re.compile('^(aya|ye|woo|ma)+$')
    answer = 0

    for e in babbling:
        if regex.match(e):
            answer += 1

    return answer

# TestCase
print(solution(["aya", "yee", "u", "maa", "wyeoo"]) == 1)
print(solution(["ayaye", "uuuma", "ye", "yemawoo", "ayaa"]) == 3)
print(solution(["ayeya", "wmaoo", "woayao"]) == 0)
```
