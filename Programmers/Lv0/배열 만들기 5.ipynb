```python
from typing import List

def solution(intStrs: List[str], k: int, s: int, l: int) -> List[int]:
    answer = []

    for num in intStrs:
        num = int(num[s:s + l])
        if num > k:
            answer.append(num)

    return answer

# TestCase
print(solution(["0123456789", "9876543210", "9999999999999"], 50000, 5, 5) == [56789, 99999])
```
