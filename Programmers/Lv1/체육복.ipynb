```python
from typing import List

def solution(n: int, lost: List[int], reserve: List[int]) -> int:
    lost_new = sorted(set(lost) - set(reserve), reverse=True)
    reserve_new = sorted(set(reserve) - set(lost), reverse=True)
    answer = n - len(lost_new)

    while lost_new and reserve_new:
        while lost_new and lost_new[-1] + 1 < reserve_new[-1]:
            del lost_new[-1]
        if lost_new and lost_new[-1] in [reserve_new[-1] - 1, reserve_new[-1] + 1]:
            answer += 1
            del lost_new[-1]
        del reserve_new[-1]

    return answer

# TestCase
print(solution(5, [2, 4], [1, 3, 5]) == 5)
print(solution(5, [2, 4], [3]) == 4)
print(solution(5, [2, 3], [3, 4, 5]) == 4)
print(solution(8, [4, 5, 7, 9], [1, 6, 8, 9]) == 7)
print(solution(5, [2, 4], [1, 3]) == 5)
print(solution(5, [4, 2], [3, 5]) == 5)
print(solution(6, [3, 5, 1], [4, 2, 6]) == 6)
```
