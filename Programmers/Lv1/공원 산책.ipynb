```python
from typing import List

def solution(park: List[str], routes: List[str]) -> List[int]:
    op_dict = {"W": (0, -1), "E": (0, 1), "N": (-1, 0), "S": (1, 0)}
    w, h = len(park[0]), len(park)
    for r, row in enumerate(park):
        for c, col in enumerate(row):
            if col == 'S':
                y, x = r, c
                break

    for route in routes:
        direct, size = route.split()
        size = int(size)
        a, b = op_dict[direct]
        a, b = a * size, b * size
        y_, x_ = y + a, x + b
        if 0 <= x_ < w and 0 <= y_ < h:
            x_min, x_max = min(x, x_), max(x, x_)
            y_min, y_max = min(y, y_), max(y, y_)

            cond1 = direct in ['W', 'E'] and 'X' not in park[y][x_min:x_max + 1]
            cond2 = direct in ['N', 'S'] and 'X' not in [park[i][x] for i in range(y_min, y_max + 1)]
            if cond1 or cond2:
                y, x = y_, x_

    return [y, x]

# TestCase
print(solution(["SOO", "OOO", "OOO"], ["E 2", "S 2", "W 1"]) == [2, 1])
print(solution(["SOO", "OXX", "OOO"], ["E 2", "S 2", "W 1"]) == [0, 1])
print(solution(["OSO", "OOO", "OXO", "OOO"], ["E 2", "S 3", "W 1"]) == [0, 0])
```
