```python
from typing import List

def solution(N: int, stages: List[int]) -> List[int]:
    n = len(stages) # 게임 이용자 수
    stage_info = {i: 0 for i in range(N + 2)}
    for i in stages:
        stage_info[i] += 1

    for i in range(1, N + 1):
        stage_info[i], n = stage_info[i] / n, n - stage_info[i]
        if n <= 0:
            break

    return sorted(range(1, N + 1), key=lambda x: (-stage_info[x], x))

# TestCase
print(solution(5, [2, 1, 2, 6, 2, 4, 3, 3]) == [3, 4, 2, 1, 5])
print(solution(4, [4, 4, 4, 4, 4]) == [4, 1, 2, 3])
```
