```python
def solution(s: str) -> int:
    str_dict = {'zero':'0', 'one':'1', 'two':'2', 'three': '3', 'four':'4', 'five':'5', 'six':'6', 'seven':'7', 'eight':'8', 'nine':'9'}
    number = ''
    trans = ''

    for char in s:
        if char.isdecimal():
            number += char
        else:
            trans += char
            if trans in str_dict:
                number += str_dict[trans]
                trans = ''

    answer = int(number)
    return answer

# TestCase
print(solution("one4seveneight") == 1478)
print(solution("23four5six7") == 234567)
print(solution("2three45sixseven") == 234567)
print(solution("123") == 123)
```
