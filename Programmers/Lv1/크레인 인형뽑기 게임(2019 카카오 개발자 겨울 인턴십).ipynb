```python
from typing import List

def solution(board: List[List[int]], moves: List[int]) -> int:
    n = len(board)
    answer = 0
    stack = []
    col_info = [n - row.count(0) for row in zip(*board)]

    for col in moves:
        col -= 1
        if not col_info[col]:
            continue

        row = n - col_info[col]
        if stack and stack[-1] == board[row][col]:
            answer += 2
            del stack[-1]
        else:
            stack.append(board[row][col])

        col_info[col] -= 1

    return answer

# TestCase
print(solution([[0, 0, 0, 0, 0], [0, 0, 1, 0, 3], [0, 2, 5, 0, 1], [4, 2, 4, 4, 2], [3, 5, 1, 3, 1]], [1, 5, 3, 5, 1, 2, 1, 4]) == 4)
```
