```python
from typing import List
import heapq

def solution(k: int, score: List[int]) -> List[int]:
    hq = []
    answer = []

    for s in score:
        heapq.heappush(hq, s)
        if len(hq) > k:
            heapq.heappop(hq)
        answer.append(min(hq))

    return answer

# TestCase
print(solution(3, [10, 100, 20, 150, 1, 100, 200]) == [10, 10, 10, 20, 20, 100, 100])
print(solution(4, [0, 300, 40, 300, 20, 70, 150, 50, 500, 1000]) == [0, 0, 0, 0, 20, 40, 70, 70, 150, 300])
```
