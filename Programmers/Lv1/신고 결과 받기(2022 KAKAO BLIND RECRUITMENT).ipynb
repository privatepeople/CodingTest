```python
from typing import List

def solution(id_list: List[str], report: List[str], k: int) -> List[int]:
    report_dict = {id_: {'number': idx, 'reporter_set': set()} for idx, id_ in enumerate(id_list)}
    result = [0] * len(id_list)

    for report_info in report:
        reporter, target = report_info.split()
        report_dict[target]['reporter_set'].add(reporter)

    for target, info in report_dict.items():
        if len(info['reporter_set']) >= k:
            for reporter in info['reporter_set']:
                result[report_dict[reporter]['number']] += 1

    return result

# TestCase
print(solution(["muzi", "frodo", "apeach", "neo"], ["muzi frodo","apeach frodo","frodo neo","muzi neo","apeach muzi"], 2) == [2, 1, 1, 0])
print(solution(["con", "ryan"], ["ryan con", "ryan con", "ryan con", "ryan con"], 3) == [0, 0])
```
