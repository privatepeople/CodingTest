```python
from collections import Counter
from operator import itemgetter

def solution(X: str, Y: str) -> str:
    x_cnt_dict, y_cnt_dict = Counter(X), Counter(Y)
    if len(x_cnt_dict) > len(y_cnt_dict):
        x_cnt_dict, y_cnt_dict = y_cnt_dict, x_cnt_dict
    answer = ''

    for i, val in sorted(x_cnt_dict.items(), key=itemgetter(0), reverse=True):
        if i in y_cnt_dict:
            if not answer and i == '0':
                return '0'
            answer += (i * min(val, y_cnt_dict[i]))

    return answer if answer else '-1'

# TestCase
print(solution("100", "2345") == "-1")
print(solution("100", "203045") == "0")
print(solution("100", "123450") == "10")
print(solution("100", "42531") == "321")
print(solution("5525", "1255") == "552")
```
