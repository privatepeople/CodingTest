```python
from typing import List

def solution(cards1: List[str], cards2: List[str], goal: List[str]) -> str:
    first = second = 0
    first_len, second_len = len(cards1), len(cards2)
    for word in goal:
        if first < first_len and word == cards1[first]:
            first += 1
        elif second < second_len and word == cards2[second]:
            second += 1
        else:
            return 'No'

    return "Yes"

# TestCase
print(solution(["i", "drink", "water"], ["want", "to"], ["i", "want", "to", "drink", "water"]) == "Yes")
print(solution(["i", "water", "drink"], ["want", "to"], ["i", "want", "to", "drink", "water"]) == "No")
```
