```python
from typing import List

def solution(survey: List[str], choices: List[int]) -> str:
    answer = ''
    index_dict = {'R': 0, 'T': 0, 'C': 0, 'F': 0, 'J': 0, 'M': 0, 'A': 0, 'N': 0}

    for i, val in enumerate(choices):
        if val >= 5:
            index_dict[survey[i][1]] += (val - 4)
        elif val <= 3:
            index_dict[survey[i][0]] += [3, 2, 1][val - 1]

    answer += 'R' if index_dict['R'] >= index_dict['T'] else 'T'
    answer += 'C' if index_dict['C'] >= index_dict['F'] else 'F'
    answer += 'J' if index_dict['J'] >= index_dict['M'] else 'M'
    answer += 'A' if index_dict['A'] >= index_dict['N'] else 'N'

    return answer

# TestCase
print(solution(["AN", "CF", "MJ", "RT", "NA"], [5, 3, 2, 7, 5]) == "TCMA")
print(solution(["TR", "RT", "TR"], [7, 1, 3]) == "RCJA")
```
