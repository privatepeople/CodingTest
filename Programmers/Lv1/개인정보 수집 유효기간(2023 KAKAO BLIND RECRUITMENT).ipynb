```python
from typing import List

def solution(today: str, terms: List[str], privacies: List[str]) -> List[int]:
    kind_month = {}
    for term in terms:
        kind, mon = term.split()
        kind_month[kind] = int(mon) * 28
    year, mon, day = map(int, today.split('.'))
    now = year * 12 * 28 + mon * 28 + day
    answer = []

    for i, value in enumerate(privacies, start=1):
        info_date, kind = value.split()
        year, mon, day = map(lambda x : int(x), info_date.split('.'))
        contract = year * 12 * 28 + mon * 28 + day + kind_month[kind] - 1

        if contract < now:
            answer.append(i)

    return answer

# TestCase
print(solution("2022.05.19", ["A 6", "B 12", "C 3"], ["2021.05.02 A", "2021.07.01 B", "2022.02.19 C", "2022.02.20 C"]) == [1, 3])
print(solution("2020.01.01", ["Z 3", "D 5"], ["2019.01.01 D", "2019.11.15 Z", "2019.08.02 D", "2019.07.01 D", "2018.12.28 Z"]) == [1, 4, 5])
```
