```python
def solution(n: int) -> int:
    num_set = set(range(2, n + 1))

    for i in range(2, int(n**0.5) + 1):
        if i in num_set:
            num_set -= set(range(i * 2, n + 1, i))

    return len(num_set)

# TestCase
print(solution(10) == 4)
print(solution(5) == 3)
```
