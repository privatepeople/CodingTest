```python
from typing import List, Iterator
from collections import defaultdict
from itertools import combinations

def sieve_eratosthenes(n: int) -> Iterator[int]:
    """n이하 자연수 중에서 소수들을 오름차순으로 반환하는 제너레이터"""
    prime_set = set(range(2, n + 1))

    for i in range(2, int(n**0.5) + 1):
        if i in prime_set:
            prime_set -= set(range(i * 2, n + 1, i))

    yield from sorted(prime_set)

def solution(nums: List[int]) -> int:
    num_dict = defaultdict(int)
    for num_comb in combinations(nums, 3):
        num_dict[sum(num_comb)] += 1

    prime_set = set(sieve_eratosthenes(max(num_dict)))
    answer = 0

    for key in num_dict:
        if key in prime_set:
            answer += num_dict[key]

    return answer

# TestCase
print(solution([1, 2, 3, 4]) == 1)
print(solution([1, 2, 7, 6, 4]) == 4)
```
