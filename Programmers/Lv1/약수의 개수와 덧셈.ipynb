```python
from math import sqrt

def solution(left: int, right: int) -> int:
    answer = 0

    for i in range(left, right + 1):
        answer += -i if sqrt(i).is_integer() else i

    return answer

print(solution(13, 17) == 43)
print(solution(24, 27) == 52)
```
