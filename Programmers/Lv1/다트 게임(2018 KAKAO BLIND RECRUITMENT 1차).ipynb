문제 출처 : https://school.programmers.co.kr/learn/courses/30/lessons/17682

```python
def solution(dartResult: str) -> int:
    nums = [0]

    for s in dartResult:
        if s == 'S':
            nums.append(0)
        elif s == 'D':
            nums[-1] **= 2
            nums.append(0)
        elif s == 'T':
            nums[-1] **= 3
            nums.append(0)
        elif s == '*':
            # 이전 값, 그 이전 값 모두 두 배 처리
            nums[-2] *= 2
            if len(nums) > 2:
                nums[-3] *= 2
        elif s == '#':
            nums[-2] *= -1
        else:
            # 자릿수 올림
            nums[-1] = nums[-1] * 10 + int(s)

    return sum(nums)

# TestCase
print(solution("1S2D*3T") == 37)
print(solution("1D2S#10S") == 9)
print(solution("1D2S0T") == 3)
print(solution("1S*2T*3S") == 23)
print(solution("1D#2S*3S") == 5)
print(solution("1T2D3D#") == -4)
print(solution("1D2S3T*") == 59)
```
