```python
from typing import List

def solution(food: List[int]) -> str:
    answer = ''

    for i,n in enumerate(food[1:], start=1):
        answer += str(i) * (n // 2)

    return answer + "0" + answer[::-1]

# TestCase
print(solution([1, 3, 4, 6]) == "1223330333221")
print(solution([1, 7, 1, 2]) == "111303111")
```
