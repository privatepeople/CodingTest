```python
from typing import List

def solution(arr: List[int]) -> List[int]:
    answer = [arr[0]]
    if len(arr) == 1:
        return answer

    for num in arr[1:]:
        if answer[-1] != num:
            answer.append(num)

    return answer

# TestCase
print(solution([1, 1, 3, 3, 0, 1, 1]) == [1, 3, 0, 1])
print(solution([4, 4, 4, 3, 3]) == [4, 3])
```
