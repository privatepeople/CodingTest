```python
def solution(s: str, n: int) -> str:
    s_dict = {}
    for i in range(65, 91):
        s_dict[chr(i)] = chr(i + n) if i + n < 91 else chr(((i + n) % 91) + 65)
    for i in range(97, 123):
        s_dict[chr(i)] = chr(i + n) if i + n < 123 else chr(((i + n) % 123) + 97)

    answer = s.translate(str.maketrans(s_dict))
    return answer

# TestCase
print(solution("AB", 1) == "BC")
print(solution("z", 1) == "a")
print(solution("a B z", 4) == "e F d")
```
