```python
from typing import List

def solution(arr1: List[List[int]], arr2: List[List[int]]) -> List[List[int]]:
    return [[a + b for a, b in zip(row_a, row_b)] for row_a, row_b in zip(arr1, arr2)]

# TestCase
print(solution([[1, 2], [2, 3]], [[3, 4], [5, 6]]) == [[4, 6], [7, 9]])
print(solution([[1], [2]], [[3], [4]]) == [[4], [6]])
```
