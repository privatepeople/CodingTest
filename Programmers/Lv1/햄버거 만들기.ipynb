```python
from typing import List

def solution(ingredient: List[int]) -> int:
    index_list = []
    answer = 0

    for i in ingredient:
        index_list.append(i)
        if len(index_list) >= 4 and index_list[-4:] == [1, 2, 3, 1]:
            answer += 1
            del index_list[-4:]

    return answer

# TestCase
print(solution([2, 1, 1, 2, 3, 1, 2, 3, 1]) == 2)
print(solution([1, 3, 2, 1, 2, 1, 3, 1, 2]) == 0)
```
