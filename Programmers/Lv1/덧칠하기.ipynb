```python
from typing import List

def solution(n: int, m: int, section: List[int]) -> int:
    answer = 1
    check = section[0] + m - 1

    for num in section:
        if num > check:
            answer += 1
            check = num + m - 1

    return answer

# TestCase
print(solution(8, 4, [2, 3, 6]) == 2)
print(solution(5, 4, [1, 3]) == 1)
print(solution(4, 1, [1, 2, 3, 4]) == 4)
```
