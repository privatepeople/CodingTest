```python
from typing import List

def solution(s: str) -> List[int]:
    answer = []
    s_dict = {}

    for idx, char in enumerate(s):
        if char not in s_dict:
            answer.append(-1)
        else:
            answer.append(idx - s_dict[char])
        s_dict[char] = idx

    return answer

# TestCase
print(solution("banana") == [-1, -1, -1, 2, 2, 2])
print(solution("foobar") == [-1, -1, 1, -1, -1, -1])
```
