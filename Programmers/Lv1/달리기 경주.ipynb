```python
from typing import List

def solution(players: List[str], callings: List[str]) -> List[str]:
    rank_info = {people: rank for rank, people in enumerate(players)}

    for calling in callings:
        calling_index = rank_info[calling]
        prev_player = players[calling_index - 1]
        players[calling_index - 1], players[calling_index] = players[calling_index], players[calling_index - 1]
        rank_info[prev_player], rank_info[calling] = rank_info[calling], rank_info[prev_player]

    return players

# TestCase
print(solution(["mumu", "soe", "poe", "kai", "mine"], ["kai", "kai", "mine", "mine"]) == ["mumu", "kai", "mine", "soe", "poe"])
```
