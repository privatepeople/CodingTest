```python
from math import sqrt, ceil

def solution(n: int) -> int:
    num = ceil(sqrt(n))
    answer = 0 if (n // num) != num else num

    for i in range(1, num):
        q, r = divmod(n, i)
        if r == 0:
            answer += (i + q)

    return answer

# 이 문제는 n = 0일 때 약수의 합이 0이라고 하는 오류가 있음
# TestCase
print(solution(12) == 28)
print(solution(5) == 6)
```
