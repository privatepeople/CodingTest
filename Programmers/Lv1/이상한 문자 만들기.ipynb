```python
def solution(s: str) -> str:
    s_ = map(lambda x: "".join([char.lower() if i & 1 else char.upper() for i, char in enumerate(x)]), s.split(" "))
    return " ".join(s_)

# TestCase
print(solution("try hello world") == "TrY HeLlO WoRlD")
```
