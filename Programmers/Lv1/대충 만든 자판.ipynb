```python
from typing import List

def solution(keymap: List[str], targets: List[str]) -> List[int]:
    answer = [0] * len(targets)
    char_dict = {}

    for number in keymap:
        for idx, char in enumerate(number, start=1):
            if char not in char_dict:
                char_dict[char] = idx
            else:
                char_dict[char] = min(char_dict[char], idx)

    for idx, target in enumerate(targets):
        count = 0
        for char in target:
            if char in char_dict:
                count += char_dict[char]
            else:
                answer[idx] = -1
                break
        else:
            answer[idx] = count

    return answer

# TestCase
print(solution(["ABACD", "BCEFD"], ["ABCD", "AABB"]) == [9, 4])
print(solution(["AA"], ["B"]) == [-1])
print(solution(["AGZ", "BSSS"], ["ASA", "BGZ"]) == [4, 6])
```
