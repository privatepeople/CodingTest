```python
from typing import List
import re

def solution(babbling: List[str]) -> int:
    p1 = re.compile(r'^(aya|ye|woo|ma)+$')
    p2 = re.compile(r'(ayaaya|yeye|woowoo|mama)+')
    answer = 0

    for word in babbling:
        if p1.match(word) and (p2.search(word) is None):
            answer += 1

    return answer

# TestCase
print(solution(["aya", "yee", "u", "maa"]) == 1)
print(solution(["ayaye", "uuu", "yeye", "yemawoo", "ayaayaa"]) == 2)
```
