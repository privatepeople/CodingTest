```python
from typing import List

def solution(wallpaper: List[str]) -> List[int]:
    width, height = len(wallpaper[0]), len(wallpaper)
    lux, luy, rdx, rdy = height, width, 0, 0

    for row_idx, row in enumerate(wallpaper):
        if "#" in row:
            lux = min(lux, row_idx)
            rdx = max(rdx, row_idx)
            luy = min(luy, row.index("#"))
            rdy = max(rdy, width - row[::-1].index("#") - 1)

    return [lux, luy, rdx + 1, rdy + 1]

# TestCase
print(solution([".#...", "..#..", "...#."]) == [0, 1, 3, 4])
print(solution(["..........", ".....#....", "......##..", "...##.....", "....#....."]) == [1, 3, 5, 8])
print(solution([".##...##.", "#..#.#..#", "#...#...#", ".#.....#.", "..#...#..", "...#.#...", "....#...."]) == [0, 0, 7, 9])
print(solution(["..", "#."]) == [1, 0, 2, 1])
```
