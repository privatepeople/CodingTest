```python
import re

def solution(new_id: str) -> str:
    new_id = new_id.lower()
    new_id = re.sub(r'[^a-z0-9_\.\-]', '', new_id)
    new_id = re.sub(r'\.{2,}', '.', new_id)
    new_id = re.sub(r'^\.{1}|\.{1}$', '', new_id)
    if not new_id:
        new_id = 'a'
    if len(new_id) >= 16:
        new_id = new_id[:15]
        if new_id[-1] == '.':
            new_id = new_id[:-1]
    new_id = new_id.ljust(3, new_id[-1])

    return new_id

# TestCase
print(solution("...!@BaT#*..y.abcdefghijklm") == "bat.y.abcdefghi")
print(solution("z-+.^.") == "z--")
print(solution("=.=") == "aaa")
print(solution("123_.def") == "123_.def")
print(solution(	"abcdefghijklmn.p") == "abcdefghijklmn")
```
