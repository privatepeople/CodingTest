```python
from typing import List

def solution(d: List[int], budget: int) -> int:
    answer = 0
    d.sort()

    for i in d:
        if budget - i >= 0:
            answer += 1
            budget -= i
        else:
            break

    return answer

# TestCase
print(solution([1, 3, 2, 5, 4], 9) == 3)
print(solution([2, 2, 3, 3], 10) == 4)
```
