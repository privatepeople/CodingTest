```python
from typing import List
from itertools import product
from operator import itemgetter

def solution(users: List[List[int]], emoticons: List[int]) -> List[int]:
    answer = []

    for discount_list in product([10, 20, 30, 40], repeat=len(emoticons)):
        service_total, prices_total, product_sales = 0, 0, list(zip(emoticons, discount_list))
        for user in users:
            prices = sum((price * (100 - sale)) // 100 for price, sale in product_sales if sale >= user[0])
            if prices >= user[1]:
                service_total += 1
            else:
                prices_total += prices

        answer.append([service_total, prices_total])

    return max(answer, key=itemgetter(0, 1))

# TestCase
print(solution([[40, 10000], [25, 10000]], [7000, 9000]) == [1, 5400])
print(solution([[40, 2900], [23, 10000], [11, 5200], [5, 5900], [40, 3100], [27, 9200], [32, 6900]], [1300, 1500, 1600, 4900]) == [4, 13860])
```
