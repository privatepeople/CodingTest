문제 출처 : https://school.programmers.co.kr/learn/courses/30/lessons/12909

```python
def solution(s: str) -> bool:
    stack = []
    for char in s:
        if char == '(':
            stack.append(char)
        else:
            if not stack:
                return False
            else:
                del stack[-1]

    return True if not stack else False

# TestCase
print(solution("()()") == True)
print(solution("(())()") == True)
print(solution(")()(") == False)
print(solution("(()(") == False)
```
