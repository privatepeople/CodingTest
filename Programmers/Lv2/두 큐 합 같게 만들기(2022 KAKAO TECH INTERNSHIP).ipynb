```python
from typing import List
from itertools import accumulate

def solution(queue1: List[int], queue2: List[int]) -> int:
    cnt = 0
    len_q = len(queue1)
    accum_list, accum_len = [0] + list(accumulate(queue1 + queue2)), 2 * len_q + 1
    object_sum = accum_list[-1] / 2
    if object_sum.is_integer():
        object_sum = int(object_sum)
    else:
        return -1

    left, right = 0, len_q
    while left <= right and right < accum_len:
        cal = accum_list[right] - accum_list[left]
        if cal == object_sum:
            return cnt
        elif cal > object_sum:
            left += 1
        else:
            right += 1
        cnt += 1

    return -1

# TestCase
print(solution([3, 2, 7, 2], [4, 6, 5, 1]) == 2)
print(solution([1, 2, 1, 2], [1, 10, 1, 2]) == 7)
print(solution([1, 1], [1, 5]) == -1)
```
