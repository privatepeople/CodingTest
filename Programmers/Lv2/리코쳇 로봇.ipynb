```python
from typing import List
from collections import deque

def solution(board: List[str]) -> int:
    board = list(map(list, board))
    r, c = len(board), len(board[0])
    start_x = start_y = None
    for x, row in enumerate(board):
        for y, col in enumerate(row):
            if col == 'R':
                start_x, start_y = x, y
        if None not in [start_x, start_y]:
            break

    board[start_x][start_y] == 'X'
    dq = deque([(start_x, start_y, 0)])
    while dq:
        x, y, move = dq.popleft()
        for a, b in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            x_, y_ = x, y
            while (0 <= x_ + a < r) and (0 <= y_ + b < c) and board[x_ + a][y_ + b] != 'D':
                x_, y_ = x_ + a, y_ + b
            if board[x_][y_] == 'G':
                return move + 1
            elif board[x_][y_] == '.':
                board[x_][y_] = 'X'
                dq.append((x_, y_, move + 1))

    return -1

# TestCase
print(solution(["...D..R", ".D.G...", "....D.D", "D....D.", "..D...."]) == 7)
print(solution([".D.R", "....", ".G..", "...D"]) == -1)
```
