```python
from collections import Counter

def solution(str1: str, str2: str) -> int:
    str1, str2 = str1.lower(), str2.lower()
    str1_list = [strs for i in range(len(str1) - 1) if (strs:= str1[i:i + 2]).isalpha()]
    str2_list = [strs for i in range(len(str2) - 1) if (strs:= str2[i:i + 2]).isalpha()]
    str1_count, str2_count = Counter(str1_list), Counter(str2_list)
    p = sum((str1_count & str2_count).values())
    q = sum((str1_count | str2_count).values())

    return int((p / q) * 65536) if q else 65536

# TestCase
print(solution("FRANCE", "french") == 16384)
print(solution("handshake", "shake hands") == 65536)
print(solution("aa1+aa2", "AAAA12") == 43690)
print(solution("E=M*C^2", "e=m*c^2") == 65536)
```
