```python
from typing import List
from collections import deque

def solution(relation: List[List[str]]) -> int:
    r, c = len(relation), len(relation[0])
    dq = deque([(i,) for i in range(c)])
    answer = 0
    res = []

    while dq:
        idx_list = dq.popleft()
        # idx_list에 있는 속성집합들로 구성된 튜플집합
        check_set = set(tuple(relation[i][j] for j in idx_list) for i in range(r))

        # 유일성 만족여부 확인
        if len(check_set) == r:
            check = set(idx_list)
            # 최소성 만족여부 확인
            for key_set in res:
                if check.issubset(key_set) or key_set.issubset(check):
                    break
            else:
                # 유일성과 최소성을 모두 만족
                answer += 1
                res.append(check)
        else:
            # 유일성을 만족안하므로 다른 열들과 복합키 구성
            for i in range(idx_list[-1] + 1, c):
                dq.append(idx_list + (i,))

    return answer

# TestCase
print(solution([["100", "ryan", "music", "2"], ["200", "apeach", "math", "2"], ["300", "tube", "computer", "3"], ["400", "con", "computer", "4"], ["500", "muzi", "music", "3"], ["600", "apeach", "music", "2"]]) == 2)
```
