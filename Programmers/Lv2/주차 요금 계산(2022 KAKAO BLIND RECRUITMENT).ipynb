```python
from typing import List
from collections import defaultdict
from math import ceil

def solution(fees: List[int], records: List[str]) -> List[int]:
    standard_time, standard_fee, unit_time, unit_fee = fees
    car_dict = {}
    time_dict = defaultdict(int)

    # 나갈 때 주차시간 계산
    for record in records:
        time, car, state = record.split()
        if car not in car_dict:
            car_dict[car] = time
        else:
            in_time = car_dict[car]
            del car_dict[car]

            cal_time = 60 * (int(time[:2]) - int(in_time[:2])) + (int(time[3:]) - int(in_time[3:]))
            time_dict[car] += cal_time

    # 23:59까지 출차하지 않은 경우 주차시간 계산
    for car in car_dict:
        in_time = car_dict[car]
        cal_time = 1439 - 60 * int(in_time[:2]) - int(in_time[3:])
        time_dict[car] += cal_time

    for car, time in time_dict.items():
        car_dict[car] = standard_fee + max(ceil((time - standard_time) / unit_time), 0) * unit_fee

    return [car[1] for car in sorted(car_dict.items(), key=lambda x: int(x[0]))]

# TestCase
print(solution([180, 5000, 10, 600], ["05:34 5961 IN", "06:00 0000 IN", "06:34 0000 OUT", "07:59 5961 OUT", "07:59 0148 IN", "18:59 0000 IN", "19:09 0148 OUT", "22:59 5961 IN", "23:00 5961 OUT"]) == [14600, 34400, 5000])
print(solution([120, 0, 60, 591], ["16:00 3961 IN", "16:00 0202 IN", "18:00 3961 OUT", "18:00 0202 OUT", "23:58 3961 IN"]) == [0, 591])
print(solution([1, 461, 1, 10], ["00:00 1234 IN"]) == [14841])
```
