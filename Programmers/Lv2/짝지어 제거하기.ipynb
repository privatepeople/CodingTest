```python
def solution(s: str) -> int:
    stack = []
    for char in s:
        if stack and char == stack[-1]:
            del stack[-1]
        else:
            stack.append(char)

    return 0 if stack else 1

# TestCase
print(solution("baabaa") == 1)
print(solution("cdcd") == 0)
```
