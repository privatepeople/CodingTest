```python
from typing import List

def solution(msg: str) -> List[int]:
    dict = {}
    for num in range(0, 26):
        alpha = chr(num + 65)
        dict[alpha] = num + 1

    # A ~ Z 사전을 생성
    answer = [0]
    value = 26
    base = ""

    for i in range(len(msg)):
        base += msg[i]
        if not base in dict:
            value += 1
            dict[base] = value

            base = msg[i]
            answer.append(dict[base])
            # 사전에 없다면 사전에 추가하고, base를 msg[i]로 변경
        else:
            answer[-1] = dict[base]
            # base가 사전에 있다면 answer에 마지막으로 추가한 숫자를 변경

    return answer

# TestCase
print(solution("KAKAO") == [11, 1, 27, 15])
print(solution("TOBEORNOTTOBEORTOBEORNOT") == [20, 15, 2, 5, 15, 18, 14, 15, 20, 27, 29, 31, 36, 30, 32, 34])
print(solution("ABABABABABABABAB") == [1, 2, 27, 29, 28, 31, 30])
```
