```python
from typing import List
from functools import reduce
from math import gcd

def solution(nums1: List[int], nums2: List[int]) -> int:
    gcd1, gcd2 = reduce(gcd, nums1), reduce(gcd, nums2)
    answer = []

    if all(each % gcd2 for each in nums1):
        answer.append(gcd2)
    if all(each % gcd1 for each in nums2):
        answer.append(gcd1)

    return max(answer) if answer else 0

# TestCase
print(solution([10, 17], [5, 20]) == 0)
print(solution([10, 20], [5, 17]) == 10)
print(solution([14, 35, 119], [18, 30, 102]) == 7)
```
