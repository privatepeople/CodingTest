```python
from typing import List

def solution(k: int, ranges: List[List[int]]) -> List[float]:
    seq = [k]
    n = 0
    while seq[-1] > 1:
        n += 1
        if seq[-1] & 1 == 1:
            seq.append(3 * seq[-1] + 1)
        else:
            seq.append(seq[-1] >> 1)
    answer = []

    for a, b in ranges:
        b += n
        if a == b:
            answer.append(0.0)
        elif a > b:
            answer.append(-1.0)
        else:
            area = 0
            for i in range(a, b):
                area += (seq[i] + seq[i + 1]) / 2
            answer.append(round(area, 1))

    return answer

# TestCase
print(solution(5, [[0, 0], [0, -1], [2, -3], [3, -3]]) == [33.0, 31.5, 0.0, -1.0])
print(solution(3, [[0, 0], [1, -2], [3, -3]]) == [47.0, 36.0, 12.0])
```
