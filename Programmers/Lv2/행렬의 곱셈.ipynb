```python
from typing import List

def solution(arr1: List[List[int]], arr2: List[List[int]]) -> List[List[int]]:
    return [[sum(a * b for a, b in zip(A_row,B_col)) for B_col in zip(*arr2)] for A_row in arr1]

# TestCase
print(solution([[1, 4], [3, 2], [4, 1]], [[3, 3], [3, 3]]) == [[15, 15], [15, 15], [15, 15]])
print(solution([[2, 3, 2], [4, 2, 4], [3, 1, 4]], [[5, 4, 3], [2, 4, 1], [3, 1, 1]]) == [[22, 22, 11], [36, 28, 18], [29, 20, 14]])
```
