```python
from collections import Counter

def dfs(s_list: list) -> bool:
    """괄호가 올바르게 되어있는지 확인하는 함수"""
    s_dict = {')': '(', '}': '{', ']': '['}
    stack = []
    for char in s_list:
        if char in [')', '}', ']']:
            if stack and stack[-1] == s_dict[char]:
                del stack[-1]
            else:
                return False
        else:
            stack.append(char)

    return False if stack else True


def solution(s: str) -> int:
    s_dict = {')': '(', '}': '{', ']': '['}
    s_count = Counter(s)
    for char in [')', '}', ']']:
        if s_count[char] != s_count[s_dict[char]]:
            return 0

    s = list(s * 2)
    idx, n = 0, len(s) // 2
    answer = 0
    while idx <= n - 1:
        if dfs(s[idx:idx + n]):
            answer += 1
            idx += 2
        else:
            idx += 1
            while idx <= n - 1 and s[idx] == s[idx - 1]:
                idx += 1

    return answer

# TestCase
print(solution("[](){}") == 3)
print(solution("}]()[{") == 2)
print(solution("[)(]") == 0)
print(solution("}}}") == 0)
```
