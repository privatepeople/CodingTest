```python
def solution(n: int) -> int:
    if n & 1 == 1:
        return 0

    a = answer = 1
    i, sums = 2, 0
    while i <= n:
        answer = a * 3 + sums * 2
        sums += a
        a = answer
        i += 2

    return answer % 1000000007

# TestCase
print(solution(4) == 11)
```
