```python
from typing import List

def hanoi(start: int, target: int, temp: int, N: int) -> List[List[int]]:
    """하노이 탑의 이동순서를 반환하는 함수

    start : 시작위치
    target : 도착위치
    temp : 중간에 이동하는 위치
    N : 옮겨야 하는 원반의 갯수
    """
    if N == 1:
        return [[start, target]]
    return hanoi(start, temp, target, N - 1) + [[start, target]] + hanoi(temp, target, start, N - 1)

def solution(n: int) -> List[List[int]]:
    answer = hanoi(1, 3, 2, n)
    return answer

# TestCase
print(solution(2) == [[1, 2], [1, 3], [2, 3]])
```
