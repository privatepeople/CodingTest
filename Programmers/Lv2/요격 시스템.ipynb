```python
from typing import List
from operator import itemgetter

def solution(targets: List[List[int]]) -> int:
    targets.sort(key=itemgetter(1), reverse=True)
    answer = 0

    while targets:
        start, end = targets.pop()
        while targets and targets[-1][0] < end:
            del targets[-1]
        answer += 1

    return answer

# TestCase
print(solution([[4, 5], [4, 8], [10, 14], [11, 13], [5, 12], [3, 7], [1, 4]]) == 3)
```
