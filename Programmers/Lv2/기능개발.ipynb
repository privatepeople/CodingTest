```python
from typing import List
from math import ceil

def solution(progresses: List[int], speeds: List[int]) -> List[int]:
    remain_day = [ceil((100 - progress) / speed) for progress, speed in zip(progresses, speeds)]
    answer = []

    start, count = remain_day[0], 0
    for day in remain_day:
        if day <= start:
            count += 1
        else:
            answer.append(count)
            start, count = day, 1

    answer.append(count)
    return answer

# TestCase
print(solution([93, 30, 55], [1, 30, 5]) == [2, 1])
print(solution([95, 90, 99, 99, 80, 99], [1, 1, 1, 1, 1, 1]) == [1, 3, 2])
```
