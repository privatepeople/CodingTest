```python
from typing import List

def solution(people: List[int], limit: int) -> int:
    answer = 0
    people.sort()
    left, right = 0, len(people) - 1

    while left <= right:
        val = people[right]
        if people[left] + people[right] <= limit:
            left += 1

        right -= 1
        answer += 1

    return answer

# TestCase
print(solution([70, 50, 80, 50], 100) == 3)
print(solution([70, 80, 50], 100) == 3)
```
