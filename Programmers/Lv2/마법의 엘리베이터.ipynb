```python
def solution(storey: int) -> int:
    answer = 0

    while storey:
        if storey <= 10:
            answer += min(11 - storey, storey)
            break
        else:
            ten, one = storey % 100, storey % 10
            answer += min(10 - one, one)
            # 35, 36, 45, 46, 50, 53, 55, 56을 확인해보면 일의 자리가 5, 십의 자리가 5인 것을 기점으로 최소 횟수가 달라짐
            if (ten >= 50 and one >= 5) or (ten < 50 and one > 5):
                storey += 10
            storey //= 10

    return answer

# TestCase
print(solution(16) == 6)
print(solution(2554) == 16)
print(solution(35) == 8)
print(solution(36) == 8)
print(solution(45) == 9)
print(solution(46) == 9)
print(solution(50) == 5)
print(solution(53) == 8)
print(solution(55) == 10)
print(solution(56) == 9)
```
