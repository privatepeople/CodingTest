```python
from typing import List
import heapq

def solution(scoville: List[int], K: int) -> int:
    answer = 0
    n =  len(scoville)
    heapq.heapify(scoville)

    while scoville:
        scovi1 = heapq.heappop(scoville)
        if scovi1 >= K:
            return answer
        else:
            if scoville:
                scovi2 = heapq.heappop(scoville)
                heapq.heappush(scoville, scovi1 + 2 * scovi2)
            else:
                return -1

        answer += 1

# TestCase
print(solution([1, 2, 3, 9, 10, 12], 7) == 2)
```
