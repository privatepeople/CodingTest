```python
from math import sqrt

def solution(k: int, d: int) -> int:
    answer = 0

    for x in range(0, d + 1, k):
        answer += int(sqrt((d**2) - (x**2)) / k) + 1

    return answer

# TestCase
print(solution(2, 4) == 6)
print(solution(1, 5) == 26)
```
