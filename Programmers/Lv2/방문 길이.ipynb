```python
def solution(dirs: str) -> int:
    answer = 0
    x, y = 0, 0
    dir_dict = {'U': (0, 1), 'D': (0, -1), 'L': (-1, 0), 'R': (1, 0)}
    loc_dict = {}

    for move in dirs:
        a, b = dir_dict[move]
        x_, y_ = x + a, y + b
        if -5 <= x_ <= 5 and -5 <= y_ <= 5:
            if (x, y, x_, y_) not in loc_dict:
                loc_dict[(x, y, x_, y_)] = True
                loc_dict[(x_, y_, x, y)] = True
                answer += 1
            x, y = x_, y_

    return answer

# TestCase
print(solution("ULURRDLLU") == 7)
print(solution("LULLLLLLU") == 7)
```
