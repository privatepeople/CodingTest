```python
from typing import List
from itertools import chain

def solution(n: int) -> List[int]:
    x = y = idx = 0
    direct = [(1, 0), (0, 1), (-1, -1)]
    traingle = [[0 for j in range(i + 1)] for i in range(n)]

    for num in range(1, ((n * (n + 1)) // 2) + 1):
        traingle[x][y] = num
        x_, y_ = x + direct[idx][0], y + direct[idx][1]
        if not ((0 <= x_ < n) and (0 <= y_ < n) and traingle[x_][y_] == 0):
            idx = (idx + 1) % 3
        x, y = x + direct[idx][0], y + direct[idx][1]

    return list(chain(*traingle))

# TestCase
print(solution(4) == [1, 2, 9, 3, 10, 8, 4, 5, 6, 7])
print(solution(5) == [1, 2, 12, 3, 13, 11, 4, 14, 15, 10, 5, 6, 7, 8, 9])
print(solution(6) == [1, 2, 15, 3, 16, 14, 4, 17, 21, 13, 5, 18, 19, 20, 12, 6, 7, 8, 9, 10, 11])
```
