```python
from typing import List
from operator import itemgetter

def solution(picks: List[int], minerals: List[str]) -> int:
    mineral_dict = {'diamond': 0, 'iron': 1, 'stone': 2}
    answer = 0
    res = []
    picks_sum, n = 5 * sum(picks), len(minerals)

    if n < picks_sum:
        q, r = divmod(n, 5)
        end = q * 5
        last_minerals = minerals[end:end + r]
        cnt_list = [0, 0, 0]

        for mineral in last_minerals:
            cnt_list[mineral_dict[mineral]] += 1
        res.append(cnt_list)
    else:
        end = picks_sum

    for idx in range(0, end, 5):
        mineral_list = minerals[idx:idx + 5]
        cnt_list = [0, 0, 0]

        for mineral in mineral_list:
            cnt_list[mineral_dict[mineral]] += 1
        res.append(cnt_list)

    res.sort(key=itemgetter(0, 1, 2))
    while res:
        d, i, s = res.pop()
        if picks[0]:
            p = d + i + s
            answer += p
            picks[0] -= 1
        elif picks[1]:
            p = 5 * d + i + s
            answer += p
            picks[1] -= 1
        elif picks[2]:
            p = 25 * d + 5 * i + s
            answer += p
            picks[2] -= 1
        else:
            break

    return answer

# TestCase
print(solution([1, 3, 2], ["diamond", "diamond", "diamond", "iron", "iron", "diamond", "iron", "stone"]) == 12)
print(solution([0, 1, 1], ["diamond", "diamond", "diamond", "diamond", "diamond", "iron", "iron", "iron", "iron", "iron", "diamond"]) == 50)
print(solution([1, 1, 0], ["iron", "iron", "iron", "iron", "iron", "diamond"]) == 6)
print(solution([1, 1, 0], ["stone", "stone", "iron", "stone", "diamond", "diamond", "diamond", "diamond", "diamond", "diamond"]) == 14)
print(solution([1, 1, 0], ["iron", "iron", "diamond", "iron", "stone", "diamond", "diamond", "diamond"]) == 12)
```
