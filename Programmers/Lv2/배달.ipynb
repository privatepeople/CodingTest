```python
from typing import List
from collections import defaultdict
import heapq

def solution(N: int, road: List[List[int]], K: int) -> int:
    answer = 0
    graph = defaultdict(list)
    dist = defaultdict(int)
    q = [(0, 1)]
    for v1, v2, weight in road:
        graph[v1].append((v2, weight))
        graph[v2].append((v1, weight))

    while q:
        time, node = heapq.heappop(q)
        if node not in dist:
            dist[node] = time
            for v, weight in graph[node]:
                alt = time + weight
                if v not in dist and alt <= K:
                    heapq.heappush(q, (alt, v))

    return len(dist.values())

# TestCase
print(solution(5, [[1, 2, 1], [2, 3, 3], [5, 2, 2], [1, 4, 2], [5, 3, 1], [5, 4, 2]], 3) == 4)
print(solution(6, [[1, 2, 1], [1, 3, 2], [2, 3, 2], [3, 4, 3], [3, 5, 2], [3, 5, 3], [5, 6, 1]], 4) == 4)
```
