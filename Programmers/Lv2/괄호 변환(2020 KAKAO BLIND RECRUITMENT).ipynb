```python
def dfs(char: str) -> bool:
    """괄호가 올바르게 되어있는지 확인하는 함수"""
    stack = []
    for i in char:
        if i == ')':
            if stack:
                del stack[-1]
            else:
                return False
        else:
            stack.append(i)

    return True if not stack else False


def solution(p: str) -> str:
    if not p:
        return ''

    idx = -1
    left = right = 0

    for char in p:
        idx += 1
        if char == '(':
            left += 1
        else:
            right += 1
        if left == right:
            break

    if idx != len(p) - 1:
        u, v = p[:idx + 1], p[idx + 1:]
    else:
        u, v = p, ''

    if dfs(u):
        return u + solution(v)
    else:
        answer = ''
        for i in u[1:-1]:
            if i == '(':
                answer += ')'
            else:
                answer += '('
        return '(' + solution(v) + ')' + answer

# TestCase
print(solution("(()())()") == "(()())()")
print(solution(")(") == "()")
print(solution("()))((()") == "()(())()")
```
