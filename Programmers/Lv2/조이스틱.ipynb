```python
def solution(name: str) -> int:
    n = len(name)
    answer = 0

    move = n - 1
    for idx in range(n):
        val = ord(name[idx]) - ord('A')
        answer += min(val, 26 - val)

        next_idx = idx + 1
        while (next_idx < n) and (name[next_idx] == 'A'):
            next_idx += 1
        distance = min(idx, n - next_idx)
        move = min(move, idx + n - next_idx + distance)

    return answer + move

# TestCase
print(solution1("JEROEN") == 56)
print(solution1("JAN") == 23)
```
