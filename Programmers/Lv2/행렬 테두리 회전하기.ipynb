```python
from typing import List

def solution(rows: int, cols: int, queries: List[List[int]]) -> List[int]:
    answer = [0] * len(queries)
    map_list = [[j for j in range(i, i + cols)] for i in range(1, rows * cols, cols)]
    direct_list = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    default_n = rows * cols

    for idx, query in enumerate(queries):
        x1, y1, x2, y2 = query[0] - 1, query[1] - 1, query[2] - 1, query[3] - 1
        x, y = x1, y1
        direct = 0
        num_min = default_n
        prev_val = map_list[x + 1][y]

        for _ in range(2 * (x2 - x1) + 2 * (y2 - y1)):
            num_min = min(num_min, prev_val)
            prev_val, map_list[x][y] = map_list[x][y], prev_val
            if not ((x1 <= (x + direct_list[direct][0]) <= x2) and (y1 <= (y + direct_list[direct][1]) <= y2)):
                direct = (direct + 1) % 4

            x, y = x + direct_list[direct][0], y + direct_list[direct][1]

        answer[idx] = num_min

    return answer

# TestCase
print(solution(6, 6, [[2, 2, 5, 4], [3, 3, 6, 6], [5, 1, 6, 3]]) == [8, 10, 25])
print(solution(3, 3, [[1, 1, 2, 2], [1, 2, 2, 3], [2, 1, 3, 2], [2, 2, 3, 3]]) == [1, 1, 5, 3])
print(solution(100, 97, [[1, 1, 100, 97]]) == [1])
```
