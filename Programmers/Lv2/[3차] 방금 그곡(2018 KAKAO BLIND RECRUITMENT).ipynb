```python
from typing import List
import re

def re_trans(matching: re.Match) -> str:
    """패턴 매칭된 문자열을 알파벳부분의 소문자로 변환하는 함수

    matching : 패턴 매칭된 문자열
    """
    return matching.group('char').lower()


def solution(m: str, musicinfos: List[str]) -> str:
    answer = '(None)'
    max_time = 0
    p = re.compile(r'(?P<char>[A-Z])#')
    m = p.sub(re_trans, m)

    for info in musicinfos:
        start, end, title, music = info.split(',')
        start, end = start.split(':'), end.split(':')
        music_time = (int(end[0]) - int(start[0])) * 60 + (int(end[1]) - int(start[1]))
        music = p.sub(re_trans, music)
        q, r = divmod(music_time, len(music))
        music = music * q + music[:r]

        if m in music and music_time > max_time:
            max_time = music_time
            answer = title

    return answer

# TestCase
print(solution("ABCDEFG", ["12:00,12:14,HELLO,CDEFGAB", "13:00,13:05,WORLD,ABCDEF"]) == "HELLO")
print(solution("CC#BCC#BCC#BCC#B", ["03:00,03:30,FOO,CC#B", "04:00,04:08,BAR,CC#BCC#BCC#B"]) == "FOO")
print(solution("ABC", ["12:00,12:14,HELLO,C#DEFGAB", "13:00,13:05,WORLD,ABCDEF"]) == "WORLD")
```
