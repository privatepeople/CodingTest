```python
from typing import List

def solution(skill: str, skill_tree: List[str]) -> int:
    answer = 0
    skill_set = set(skill)

    for skilltree in skill_tree:
        idx = 0
        for s in skilltree:
            if s in skill_set:
                if s != skill[idx]:
                    break
                idx += 1
        else:
            answer += 1

    return answer

# TestCase
print(solution("CBD", ["BACDE", "CBADF", "AECB", "BDA"]) == 2)
```
