```python
from typing import List
from collections import defaultdict
from operator import itemgetter

def time_trans(times: str) -> int:
    """HH:MM형식의 시간을 자연수로 변환하는 함수"""
    return 60 * int(times[:2]) + int(times[3:])


def solution(book_time: List[List[str]]) -> int:
  # imos법을 통한 풀이
    time_dict = defaultdict(int)
    current_room_cnt = 0
    answer = 0

    for start, end in book_time:
        start, end = time_trans(start), time_trans(end)
        time_dict[start] += 1
        time_dict[end + 10] -= 1

    for times, cnt in sorted(time_dict.items(), key=itemgetter(0, 1)):
        current_room_cnt += cnt
        answer = max(answer, current_room_cnt)

    return answer

# TestCase
print(solution([["15:00", "17:00"], ["16:40", "18:20"], ["14:20", "15:20"], ["14:10", "19:20"], ["18:20", "21:20"]]) == 3)
print(solution([["09:10", "10:10"], ["10:20", "12:20"]]) == 1)
print(solution([["10:20", "12:30"], ["10:20", "12:30"], ["10:20", "12:30"]]) == 3)
```
