문제 출처 : https://school.programmers.co.kr/learn/courses/30/lessons/17687

```python
def solution(n: int, t: int, m: int, p: int) -> str:
    answer, notation = '01', "0123456789ABCDEF"
    num = 2

    while len(answer) < m * t:
        val, num_str = num, ''
        while val > 0:
            val, r = divmod(val, n)
            num_str += notation[r]

        answer += num_str[::-1]
        num += 1

    start, end, inter = p - 1, (p - 1) + (t - 1) * m + 1, m
    return answer[start:end:inter]

# TestCase
print(solution(2, 4, 2, 1) == "0111")
print(solution(16, 16, 2, 1) == "02468ACE11111111")
print(solution(16, 16, 2, 2) == "13579BDF01234567")
```
