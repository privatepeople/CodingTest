```python
from typing import List

def solution(cap: int, n: int, deliveries: List[int], pickups: List[int]) -> int:
    answer = 0
    deliver_cap = pickup_cap = 0

    # 뒤에서부터 deliver_cap과 pickup_cap 계산을 통해 정답 계산
    for i in range(n - 1, -1, -1):
        cnt = 0
        deliver_cap -= deliveries[i]
        pickup_cap -= pickups[i]
        # deliver_cap과 pickup_cap이 둘 다 음이 아닌 정수가 될 때 까지 왕복횟수 count
        # 계산 후 남은 deliver_cap과 pickup_cap의 양수값은 다음 deliveries[i]와 pickups[i]만큼 뺄 때 반영
        while deliver_cap < 0 or pickup_cap < 0:
            deliver_cap += cap
            pickup_cap += cap
            cnt += 1

        answer += (i + 1) * 2 * cnt

    return answer

# TestCase
print(solution(4, 5, [1, 0, 3, 1, 2], [0, 3, 0, 4, 0]) == 16)
print(solution(2, 7, [1, 0, 2, 0, 1, 0, 2], [0, 2, 0, 1, 0, 2, 0]) == 30)
```
