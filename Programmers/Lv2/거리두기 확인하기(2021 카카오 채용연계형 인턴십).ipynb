```python
from typing import List

def room_checking(room: List[str]) -> int:
    """거리두기 여부를 확인하는 함수"""
    for i in range(5):
        for j in range(5):
            # 사람이 앉아있으면 상하좌우에 다른 사람이 있는지 확인
            if room[i][j] == 'P':
                for a, b in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
                    x, y = i + a, j + b
                    if 0 <= x < 5 and 0 <= y < 5 and room[x][y] == 'P':
                        return 0
            # 빈 테이블이면 상하좌우에 앉은 사람의 수가 2명 미만인지 확인
            elif room[i][j] == 'O':
                cnt = 0
                for a, b in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
                    x, y = i + a, j + b
                    if 0 <= x < 5 and 0 <= y < 5 and room[x][y] == 'P':
                        cnt += 1
                        if cnt >= 2:
                            return 0

    return 1


def solution(places: List[List[str]]) -> List[int]:
    answer = [room_checking(room) for room in places]
    return answer

# TestCase
print(solution([["POOOP", "OXXOX", "OPXPX", "OOXOX", "POXXP"], ["POOPX", "OXPXP", "PXXXO", "OXXXO", "OOOPP"], ["PXOPX", "OXOXP", "OXPOX", "OXXOP", "PXPOX"], ["OOOXX", "XOOOX", "OOOXX", "OXOOX", "OOOOO"], ["PXPXP", "XPXPX", "PXPXP", "XPXPX", "PXPXP"]]) == [1, 0, 1, 1, 1])
```
