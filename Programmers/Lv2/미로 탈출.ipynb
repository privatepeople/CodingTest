```python
from typing import List
from collections import deque

def bfs(board: List[List[str]], start_x: int, start_y: int, r: int, c: int) -> int:
    """시작지점에서 도착지점까지의 최단거리를 반환하는 함수

    board : 지도
    start_x : 시작위치의 x좌표
    start_y : 시작위치의 y좌표
    r : 지도의 세로길이
    c : 지도의 가로길이
    """
    dq = deque([(start_x, start_y, 0)])
    while dq:
        x, y, move = dq.popleft()
        for a, b in [(0, 1), (1, 0), (0, -1), (-1, 0)]:
            x_, y_ = x + a, y + b
            if 0 <= x_ < r and 0 <= y_ < c:
                if board[x_][y_] == 'E':
                    return move + 1
                elif board[x_][y_] == 'O':
                    board[x_][y_] = 'X'
                    dq.append((x_, y_, move + 1))

    return -1

def solution(maps: List[str]) -> int:
    r, c = len(maps), len(maps[0])
    start_x = start_y = end_x = end_y = lever_x = lever_y = None
    for row, x in enumerate(maps):
        for col, y in enumerate(x):
            if y == 'S':
                start_x, start_y = row, col
            elif y == 'L':
                lever_x, lever_y = row, col
            elif y == 'E':
                end_x, end_y = row, col
        if None not in [start_x, start_y, end_x, end_y, lever_x, lever_y]:
            break

    maps_copy = list(map(list, maps))
    maps_copy[start_x][start_y] = 'X'
    maps_copy[end_x][end_y] = 'O'
    maps_copy[lever_x][lever_y] = 'E'

    distance1 = bfs(maps_copy, start_x, start_y, r, c)
    if distance1 == -1:
        return -1

    maps = list(map(list, maps))
    maps[start_x][start_y] = 'O'
    maps[lever_x][lever_y] = 'X'

    distance2 = bfs(maps, lever_x, lever_y, r, c)
    return distance1 + distance2 if distance2 != -1 else -1

# TestCase
print(solution(["SOOOL", "XXXXO", "OOOOO", "OXXXX", "OOOOE"]) == 16)
print(solution(["LOOXS", "OOOOX", "OOOOO", "OOOOO", "EOOOO"]) == -1)
```
