```python
from typing import List
import heapq

def solution(n: int, k: int, enemy: List[int]) -> int:
    enemy_len = len(enemy)
    if k >= enemy_len:
        return enemy_len

    hq = enemy[:k]
    heapq.heapify(hq)

    for idx, num in enumerate(enemy[k:], start=k):
        val = heapq.heappushpop(hq, num)
        n -= val
        if n < 0:
            return idx

    return enemy_len

# TestCase
print(solution(7, 3, [4, 2, 4, 5, 3, 3, 1]) == 5)
print(solution(2, 4, [3, 3, 3, 3]) == 4)
```
