```python
from typing import List
from collections import deque

def solution(maps: List[str]) -> List[int]:
    maps = list(map(list, maps))
    r, c = len(maps), len(maps[0])
    answer = []

    for i in range(r):
        for j in range(c):
            if maps[i][j] != 'X':
                res = int(maps[i][j])
                maps[i][j] = 'X'
                dq = deque([(i, j)])

                while dq:
                    x, y = dq.popleft()
                    for a, b in [(0, 1), (-1, 0), (0, -1), (1, 0)]:
                        x_, y_ = x + a, y + b
                        if 0 <= x_ < r and 0 <= y_ < c and maps[x_][y_] != 'X':
                            res += int(maps[x_][y_])
                            dq.append((x_, y_))
                            maps[x_][y_] = 'X'

                answer.append(res)

    return sorted(answer) if answer else [-1]

# TestCase
print(solution(["X591X", "X1X5X", "X231X", "1XXX1"]) == [1, 1, 27])
print(solution(["XXX", "XXX", "XXX"]) == [-1])
```
