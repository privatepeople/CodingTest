```python
from typing import List
from operator import itemgetter

def solution(plans: List[List[str]]) -> List[str]:
    plans.sort(key=itemgetter(1))
    answer = []
    stack = []

    for subject, start, times in plans:
        h, m = map(int, start.split(':'))
        start = 60 * h + m
        times = int(times)

        if stack:
            prev_subject, prev_start, prev_time = stack[-1]
            remain_time = start - prev_start

            while stack and stack[-1][2] <= remain_time:
                name, last_start, last_time = stack.pop()
                answer.append(name)
                remain_time -= last_time

            if stack:
                stack[-1] = [stack[-1][0], start, stack[-1][2] - remain_time]
        stack.append([subject, start, times])

    answer.extend([s for s, _, _ in stack[::-1]])
    return answer

# TestCase
print(solution([["korean", "11:40", "30"], ["english", "12:10", "20"], ["math", "12:30", "40"]]) == ["korean", "english", "math"])
print(solution([["science", "12:40", "50"], ["music", "12:20", "40"], ["history", "14:00", "30"], ["computer", "12:30", "100"]]) == ["science", "history", "computer", "music"])
print(solution([["aaa", "12:00", "20"], ["bbb", "12:10", "30"], ["ccc", "12:40", "10"]]) == ["bbb", "ccc", "aaa"])
```
