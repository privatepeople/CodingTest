```python
from typing import List
from itertools import accumulate

def solution(sequence: List[int], k: int) -> List[int]:
    seq_accum = [0] + list(accumulate(sequence))
    left, right, n = 0, 1, len(seq_accum)
    answer = [left, n]

    while right < n:
        cal = seq_accum[right] - seq_accum[left]
        if cal == k:
            if right - left < answer[1] - answer[0]:
                answer = [left, right]
            elif right - left == answer[1] - answer[0] and left < answer[0]:
                answer = [left, right]
            left, right = left + 1, right + 1
        elif  cal > k:
            left += 1
        else:
            right += 1

    return [answer[0], answer[1] - 1]

# TestCase
print(solution([1, 2, 3, 4, 5], 7) == [2, 3])
print(solution([1, 1, 1, 2, 3, 4, 5], 5) == [6, 6])
print(solution([2, 2, 2, 2, 2], 6) == [0, 2])
```
