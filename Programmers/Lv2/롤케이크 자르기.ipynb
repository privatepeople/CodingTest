```python
from typing import List
from collections import defaultdict, Counter

def solution(topping: List[int]) -> int:
    if len(set(topping)) == 1:
        return len(topping) - 1

    left_dict, right_dict = defaultdict(int), Counter(topping)
    answer = 0

    for val in topping[:-1]:
        left_dict[val] += 1
        right_dict[val] -= 1
        if right_dict[val] == 0:
            del right_dict[val]

        left_len, right_len = len(left_dict), len(right_dict)
        if left_len == right_len:
            answer += 1
        elif left_len > right_len:
            break

    return answer

# TestCase
print(solution([1, 2, 1, 3, 1, 4, 1, 2]) == 2)
print(solution([1, 2, 3, 1, 4]) == 0)
```
