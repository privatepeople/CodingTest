```python
from typing import List
from collections import Counter

def solution(weights: List[int]) -> int:
    weights_count = Counter(weights)
    answer = 0

    for p, cnt in weights_count.items():
        num_list = [2 * p]
        if not (p % 3):
            num_list.append((4 * p) // 3)
        if not (p % 2):
            num_list.append((3 * p) // 2)

        answer += (cnt * (cnt - 1)) // 2
        for num in num_list:
            answer += (cnt * weights_count[num])

    return answer

# TestCase
print(solution([100, 180, 360, 100, 270]) == 4)
```
