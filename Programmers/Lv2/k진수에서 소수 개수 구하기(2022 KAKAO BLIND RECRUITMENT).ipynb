```python
import re

def solution(n: int, k: int) -> int:
    answer = 0
    s = ''
    while n:
        n, r = divmod(n, k)
        s += str(r)
    s = s[::-1]
    num_list = re.findall(r'[1-9]+', s)
    answer_set = set()
    answer = 0

    for num in num_list:
        num = int(num)
        if num in answer_set:
            answer += 1
            continue

        if num != 1:
            for i in range(2, int(num**0.5) + 1):
                if not (num % i):
                    break
            else:
                answer_set.add(num)
                answer += 1

    return answer

# TestCase
print(solution(437674, 3) == 3)
print(solution(110011, 10) == 2)
```
