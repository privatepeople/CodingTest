```python
from typing import List

def solution(n: int, words: List[str]) -> List[int]:
    used_words = set( )   # 이미 사용한 단어를 저장하는 set
    prev_word = words[0][0]  # 이전 단어의 마지막 글자

    for i, word in enumerate(words):
        # 이미 사용한 단어거나 첫 글자가 이전 단어와 일치하지 않으면 탈락
        if word in used_words or word[0] != prev_word:
            q, r = divmod(i, n)
            return [r + 1, q + 1]
        used_words.add(word)  # 사용한 단어 추가
        prev_word = word[-1]  # 이전 단어의 마지막 글자 업데이트

    return [0, 0]

# TestCase
print(solution(3, ["tank", "kick", "know", "wheel", "land", "dream", "mother", "robot", "tank"]) == [3, 3])
print(solution(5, ["hello", "observe", "effect", "take", "either", "recognize", "encourage", "ensure", "establish", "hang", "gather", "refer", "reference", "estimate", "executive"]) == [0, 0])
print(solution(2, ["hello", "one", "even", "never", "now", "world", "draw"]) == [1, 3])
```
