```python
from typing import List
from collections import deque

def solution(k: int, dungeons: List[List[int]]) -> int:
    n = len(dungeons)
    answer = 0
    dq = deque()
    for idx, (min_require_p, consum_p) in enumerate(dungeons):
        if k >= min_require_p:
            dq.append([set([idx]), k - consum_p])

    while dq:
        dungeon_set, remain_k = dq.popleft()
        for i in range(n):
            if remain_k >= dungeons[i][0]:
                if i not in dungeon_set:
                    dq.append([dungeon_set | set([i]), remain_k - dungeons[i][1]])
            else:
                answer = max(answer, len(dungeon_set))

    return answer

# TestCase
print(solution(80, [[80, 20], [50, 40], [30, 10]]) == 3)
```
