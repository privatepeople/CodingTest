```python
from typing import List, Tuple
import re

def file_compile(file: str, head: re.Pattern, number: re.Pattern) -> Tuple[str, int]:
    """파일을 정렬하기 위한 키함수

    file : 해당 파일
    head : HEAD부분을 추출하는 패턴
    number: NUMBER부분을 추출하는 패턴
    """
    return (head.match(file).group().lower(), int(number.search(file).group()))


def solution(files: List[str]) -> List[str]:
    head, number = re.compile(r'\D+'), re.compile(r'\d{1,5}')
    return sorted(files, key=lambda file: file_compile(file, head, number))

# TestCase
print(solution(["img12.png", "img10.png", "img02.png", "img1.png", "IMG01.GIF", "img2.JPG"]) == ["img1.png", "IMG01.GIF", "img02.png", "img2.JPG", "img10.png", "img12.png"])
print(solution(["F-5 Freedom Fighter", "B-50 Superfortress", "A-10 Thunderbolt II", "F-14 Tomcat"]) == ["A-10 Thunderbolt II", "B-50 Superfortress", "F-5 Freedom Fighter", "F-14 Tomcat"])
```
