```python
from typing import List
from collections import Counter
from math import inf

def disjoint_find(x: int, parents: List[int]) -> int:
    """루트노드를 찾는 함수

    x : 노드 번호
    parents : 각 노드의 루트 노드들을 모아놓은 리스트
    """
    if x != parents[x]:
        parents[x] = disjoint_find(parents[x], parents)
    return parents[x]


def disjoint_union(x: int, y: int, parents: List[int], rank: List[int]) -> None:
    """루트노드를 하나로 합치는 함수

    x : 노드 번호
    y : 노드 번호
    parents : 각 노드의 루트 노드들을 모아놓은 리스트
    rank : 트리의 높이
    """
    x, y = disjoint_find(x, parents), disjoint_find(y, parents)
    if rank[x] > rank[y]:
        parents[y] = x
    else:
        parents[x] = y
    if rank[x] == rank[y]:
        rank[y] += 1
    return None


def solution(n: int, wires: List[List[int]]) -> int:
    if n in [2, 3]:
        return n - 2

    answer = inf
    e = len(wires)
    for i in range(e):
        wire =  [wires[p] for p in range(e) if p != i]
        parents = [i for i in range(n + 1)]
        rank = [0] * (n + 1)

        for v1, v2 in wire:
            disjoint_union(v1, v2, parents, rank)
        for v in range(1, n + 1):
            disjoint_find(v, parents)
        parents_count = Counter(parents[1:]).most_common()
        answer = min(answer, parents_count[0][1] - parents_count[1][1])

    return answer

# TestCase
print(solution(9, [[1, 3], [2, 3], [3, 4], [4, 5], [4, 6], [4, 7], [7, 8], [7, 9]]) == 3)
print(solution(4, [[1, 2], [2, 3], [3, 4]]) == 0)
print(solution(7, [[1, 2], [2, 7], [3, 7], [3, 4], [4, 5], [6, 7]]) == 1)
```
