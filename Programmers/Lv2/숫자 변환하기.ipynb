```python
import math

def solution(x: int, y: int, n: int) -> int:
    if x == y:
        return 0
    elif x > y:
        return -1
    else:
        y_dict = {num: math.inf for num in range(x, y + 1)}
        y_dict[x] = 0

        for idx in range(x, y):
            for num in [idx + n, idx * 2, idx * 3]:
                if num <= y:
                    y_dict[num] = min(y_dict[num], y_dict[idx] + 1)

    return -1 if math.isinf(y_dict[y]) else y_dict[y]

# TestCase
print(solution(10, 40, 5) == 2)
print(solution(10, 40, 30) == 1)
print(solution(2, 5, 4) == -1)
```
