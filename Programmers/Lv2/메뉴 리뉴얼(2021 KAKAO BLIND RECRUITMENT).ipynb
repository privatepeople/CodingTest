```python
from typing import List
from itertools import combinations
from collections import Counter

def solution(orders: List[str], course: List[int]) ->: List[str]:
    result = []

    for i in course:
        combination_list = []
        # orders의 각 order에 대해 i갯수로 만들 수 있는 조합을 더하기
        for order in orders:
            combination_list += combinations(sorted(order), i)

        combination_count = Counter(combination_list).most_common()
        if combination_count and combination_count[0][1] >= 2:
            result += [''.join(key) for key, value in combination_count if value == combination_count[0][1]]

    return sorted(result)

# TestCase
print(solution(["ABCFG", "AC", "CDE", "ACDE", "BCFG", "ACDEH"], [2, 3, 4]) == ["AC", "ACDE", "BCFG", "CDE"])
print(solution(["ABCDE", "AB", "CD", "ADE", "XYZ", "XYZ", "ACD"], [2, 3, 5]) == ["ACD", "AD", "ADE", "CD", "XYZ"])
print(solution(["XYZ", "XWY", "WXA"], [2, 3, 4]) == ["WX", "XY"])
```
