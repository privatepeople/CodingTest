```python
def solution(number: str, k: int) -> str:
    stack = []
    for num in number:
        while stack and stack[-1] < num and k > 0:
            k -= 1
            del stack[-1]
        stack.append(num)

    if k:
        stack = stack[:-k]

    return ''.join(stack)

# TestCase
print(solution("1924", 2) == "94")
print(solution("1231234", 3) == "3234")
print(solution("4177252841", 4) == "775841")
```
