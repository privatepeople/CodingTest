```python
def solution(s: str) -> str:
    s_list = sorted(s.split(), key=int)
    return s_list[0] + ' ' + s_list[-1]

# TestCase
print(solution("1 2 3 4") == "1 4")
print(solution("-1 -2 -3 -4") == "-4 -1")
print(solution("-1 -1") == "-1 -1")
```
