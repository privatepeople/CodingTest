```python
from typing import Iterator
from itertools import permutations

def eratosthenes(num: int) -> Iterator[int]:
    """n이하 자연수 중에서 소수들을 오름차순으로 반환하는 제너레이터"""
    prime_set = set(range(2, num + 1))

    for i in range(2, int(num**0.5) + 1):
        if i in prime_set:
            prime_set -= set(range(i * 2, num + 1, i))

    yield from sorted(prime_set)

def solution(numbers: str) -> int:
    max_num = int(''.join(sorted(numbers, reverse=True)))
    prime_set = set(eratosthenes(max_num))
    num_set = set()
    answer = 0

    for i in range(1, len(numbers) + 1):
        for num_perm in permutations(numbers, i):
            num = int("".join(num_perm))
            num_set.add(num)

    for num in num_set:
        if num in prime_set:
            answer += 1

    return answer

# TestCase
print(solution("17") == 3)
print(solution("011") == 2)
```
