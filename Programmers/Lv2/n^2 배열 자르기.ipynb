```python
from typing import List

def solution(n: int, left: int, right: int) -> List[int]:
    return [max(i // n, i % n) + 1 for i in range(left, right + 1)]

# TestCase
print(solution(3, 2, 5) == [3, 2, 2, 3])
print(solution(4, 7, 14) == [4, 3, 3, 3, 4, 4, 4, 4])
```
