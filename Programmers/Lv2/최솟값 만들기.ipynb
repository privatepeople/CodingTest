```python
from typing import List

def solution(A: List[int], B: List[int]) -> int:
    return sum(x * y for x, y in zip(sorted(A), sorted(B, reverse=True)))

# TestCase
print(solution([1, 4, 2], [5, 4, 4]) == 29)
print(solution([1, 2], [3, 4]) == 10)
```
