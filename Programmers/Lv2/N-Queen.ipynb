```python
from typing import List, Set

def nqueen(col_list: List[int], remain_col: Set[int]) -> int:
    """n x n에서 퀸이 서로 위협당하지 않게 배치하는 방법의 수를 반환하는 함수

    col_list : 현재까지 배치된 퀸들의 열의 위치 리스트
    remain_col : 남은 퀸들을 배치할 수 있는 열의 위치 집합
    """
    answer = 0
    if not remain_col:
        answer += 1
    else:
        n = len(col_list)
        for i in remain_col:
            for row, col in enumerate(col_list):
                if abs(n - row) == abs(col - i):
                    break
            else:
                answer += nqueen(col_list + [i], remain_col - set((i,)))

    return answer

def solution(n: int) -> int:
    if n == 1:
        return 1

    answer = 0
    total_col = set(range(n))
    for i in range(n):
        answer += nqueen([i], total_col - set((i,)))

    return answer

# TestCase
print(solution(4) == 2)
print(solution(1) == 1)
print(solution(2) == 0)
print(solution(3) == 0)
```
