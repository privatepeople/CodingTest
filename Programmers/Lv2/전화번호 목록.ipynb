```python
from typing import List
from collections import defaultdict

class TrieNode:
    """트라이 자료구조의 노드를 나타내는 클래스"""
    def __init__(self) -> None:
        """클래스 생성자"""
        self.word = False
        self.children = defaultdict(TrieNode)


class Trie:
    """트라이 자료구조를 나타내는 클래스"""
    def __init__(self) -> None:
        """클래스 생성자"""
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        """단어를 삽입하는 메소드"""
        node = self.root
        for char in word:
            node = node.children[char]
        node.word = True
        return None

    def search(self, word: str) -> bool:
        """어떤 단어가 트라이 자료구조에 있는지 확인하는 메소드"""
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.word

    def startswith(self, word: str) -> bool:
        """해당 접두사를 갖는 단어가 있는지 확인하는 메소드"""
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return True


def solution(phone_book: List[str]) -> bool:
    trie = Trie()
    phone_book = sorted(phone_book, reverse=True)

    for word in phone_book:
        if trie.startswith(word):
            return False
        trie.insert(word)

    return True

# TestCase
print(solution(["119", "97674223", "1195524421"]) == False)
print(solution(["123", "456", "789"]) == True)
print(solution(["12", "123", "1235", "567", "88"]) == False)
```
