```python
from typing import List
from collections import Counter
from functools import reduce

def solution(clothes: List[List[str]]) -> int:
    # 각 의상 종류들의 갯수에 안 입는다는 선택도 추가하여 다 곱함
    # 그리고 모든 종류의 의상들을 안 입은 1가지를 제외하면 정답
    clothes_count = Counter([kind for name, kind in clothes]).values()
    return reduce(lambda x, y: x * (y + 1), clothes_count, 1) - 1

# TestCase
print(solution([["yellow_hat", "headgear"], ["blue_sunglasses", "eyewear"], ["green_turban", "headgear"]]) == 5)
print(solution([["crow_mask", "face"], ["blue_sunglasses", "face"], ["smoky_makeup", "face"]]) == 3)
```
