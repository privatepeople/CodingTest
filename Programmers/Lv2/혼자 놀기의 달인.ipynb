```python
from typing import List
from collections import Counter

def disjoint_find(x: int, parents: List[int]) -> int:
    """루트노드를 찾는 함수

    x : 노드 번호
    parents : 각 노드의 루트 노드들을 모아놓은 리스트
    """
    if x != parents[x]:
        parents[x] = disjoint_find(parents[x], parents)
    return parents[x]


def disjoint_union(x: int, y: int, parents: List[int], rank: List[int]) -> None:
    """루트노드를 하나로 합치는 함수

    x : 노드 번호
    y : 노드 번호
    parents : 각 노드의 루트 노드들을 모아놓은 리스트
    rank : 트리의 높이
    """
    x, y = disjoint_find(x, parents), disjoint_find(y, parents)
    if rank[x] > rank[y]:
        parents[y] = x
    else:
        parents[x] = y
    if rank[x] == rank[y]:
        rank[y] += 1
    return None


def solution(cards: List[int]) -> int:
    card_len = len(cards)
    parents = [i for i in range(card_len + 1)]
    rank = [0] * (card_len + 1)
    for val in cards:
        disjoint_union(val, cards[val - 1], parents, rank)
    for val in cards:
        disjoint_find(val, parents)

    count_list = Counter(parents[1:]).most_common()

    return 0 if len(count_list) <= 1 else count_list[0][1] * count_list[1][1]

# TestCase
print(solution([8, 6, 3, 7, 2, 5, 1, 4]) == 12)
```
