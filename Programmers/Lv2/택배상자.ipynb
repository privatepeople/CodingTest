```python
from typing import List

def solution(order: List[int]) -> int:
    answer = 0
    stack = []
    for i in range(1, len(order) + 1):
        if i != order[answer]:
            stack.append(i)
        else:
            answer += 1
            while stack:
                if stack[-1] == order[answer]:
                    del stack[-1]
                    answer += 1
                else:
                    break

    return answer

# TestCase
print(solution([4, 3, 1, 2, 5]) == 2)
print(solution([5, 4, 3, 2, 1]) == 5)
```
