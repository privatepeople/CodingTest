```python
from typing import List

def solution(s: str) -> List[int]:
    number = remove_zero = 0

    while s != '1':
        zero_cnt = s.count('0')
        s = bin(len(s) - zero_cnt)[2:]
        remove_zero += zero_cnt
        number += 1

    return [number, remove_zero]

# TestCase
print(solution("110010101001") == [3, 8])
print(solution("01110") == [3, 3])
print(solution("1111111") == [4, 1])
```
