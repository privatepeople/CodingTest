```python
from typing import List

def solution(elements: List[int]) -> int:
    answer = set()
    n = len(elements)

    for i in range(n):
        ssum = 0
        for j in range(n):
            ssum += elements[(i + j) % n]
            answer.add(ssum)

    return len(answer)

# TestCase
print(solution([7, 9, 1, 1, 4]) == 18)
```
