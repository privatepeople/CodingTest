```python
from typing import List
from math import inf

def solution(m: int, n: int, startX: int, startY: int, balls: List[List[int]]) -> List[int]:
    answer = [0] * len(balls)

    # 직사각형 꼭짓점에 맞는 경우는 삼각형의 결정조건에 따라 최솟값 불가능
    # 직사각형 각 변에 대해 점을 대칭이동 시켜서 거리계산하면 각 변마다 4개의 최솟값이 생성
    # 4개의 최솟값 중 맞춰야 하는 공이 경로에 있으면 무한대로 변경
    # 이렇게 계산한 4개의 최솟값 중 최솟값이 정답
    for idx, (x, y) in enumerate(balls):
        val1 = (startX + x)**2 + (startY - y)**2
        val2 = (startX - x)**2 + (startY + y)**2
        val3 = (2 * m - startX - x)**2 + (startY - y)**2
        val4 = (startX - x)**2 + (2 * n - startY - y)**2
        if startY == y:
            if x < startX:
                val1 = inf
            else:
                val3 = inf
        if startX == x:
            if y < startY:
                val2 = inf
            else:
                val4 =  inf

        answer[idx] = min(val1, val2, val3, val4)

    return answer

# TestCase
print(solution(10, 10, 3, 7, [[7, 7], [2, 7], [7, 3]]) == [52, 37, 116])
```
