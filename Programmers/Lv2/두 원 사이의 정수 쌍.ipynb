```python
from math import sqrt, ceil, floor

def solution(r1: int, r2: int) -> int:
    answer = (r2 - r1 + 1)

    for i in range(1, r1):
        answer += (floor(sqrt((r2**2) - (i**2))) - ceil(sqrt((r1**2) - (i**2))) + 1)
    for i in range(r1, r2):
        answer += floor(sqrt((r2**2) - (i**2)))

    return answer * 4

# TestCase
print(solution(2, 3) == 20)
```
