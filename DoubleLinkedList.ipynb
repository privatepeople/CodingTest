```python
class Node:
    def __init__(self, item=None, prev=None, nexts=None):
        self.item = item
        self.prev = prev
        self.nexts = nexts

class DoubleLinkedList:
    def __init__(self):
        self.head = Node(item=None)
        self.tail = self.head
        self.__size = 0

    def __len__(self) -> int:
        return self.__size

    def __getitem__(self, idx) -> any:
        return self.get(idx)

    def check_size(self, message):
        if self.__size == 0:
            raise IndexError(message)

    def check_index(self, idx, min_idx=0, max_idx=None):
        if not isinstance(idx, int):
            raise TypeError(f"The idx argument {idx} must be a integer")
        elif (idx < min_idx) or (max_idx is not None and idx >= max_idx):
            raise IndexError(f"Index {idx} is out of bounds. Must be between {min_idx} and {self.__size - 1}")

    def append(self, item) -> None:
        new_node = Node(item=item, prev=self.tail)
        self.tail.nexts = new_node
        self.tail = new_node

        self.__size += 1

        return None

    def appendleft(self, item) -> None:
        new_node = Node(item=item, prev=self.head, nexts=self.head.nexts)

        if self.head.nexts is not None:
            self.head.nexts.prev = new_node
        else:
            self.tail = new_node

        self.head.nexts = new_node

        self.__size += 1

        return None

    def pop(self):
        self.check_size("pop from empty DoubleLinkedList")

        self.__size -= 1

        value = self.tail.item
        self.tail = self.tail.prev
        self.tail.nexts = None

        return value

    def popleft(self):
        self.check_size("popleft from empty DoubleLinkedList")

        self.__size -= 1

        result = self.head.nexts
        value = result.item

        if result.nexts is not None:
            self.head.nexts, result.nexts.prev = result.nexts, self.head
        else:
            self.head.nexts = None

        return value

    def insert(self, idx, item) -> None:
        self.check_index(idx=idx)

        idx = min(idx, self.__size)

        node = self.head
        new_node = Node(item=item)

        for _ in range(idx):
            node = node.nexts

        if node.nexts is not None:
            node.nexts.prev = new_node
            new_node.prev, new_node.nexts = node, node.nexts
            node.nexts = new_node
        else:
            new_node.prev = node
            node.nexts = new_node
            self.tail = new_node

        self.__size += 1

        return None

    def remove(self, idx) -> None:
        self.check_size("remove from empty DoubleLinkedList")
        self.check_index(idx=idx, max_idx=self.__size)

        node = self.head

        for _ in range(idx):
            node = node.nexts

        node.nexts = node.nexts.nexts
        if node.nexts is not None:
            node.nexts.prev = node
        else:
            self.tail = node

        self.__size -= 1

        return None

    def get(self, idx) -> any:
        self.check_size(f"get from empty DoubleLinkedList")
        self.check_index(idx=idx, max_idx=self.__size)

        node = self.head.nexts

        for _ in range(idx):
            node = node.nexts

        return node.item
```

```python
linked_list = DoubleLinkedList()

for i in range(1, 11):
    linked_list.append(i)

for i in range(10):
    print(linked_list.pop())

print(f"Size of DoubleLinkedList: {len(linked_list)}")
print(f"Next of DoubleLinkedList's Head is None => {linked_list.head.nexts is None}")
```
